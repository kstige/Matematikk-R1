# Derivasjonsregler

:class: utbytte, dropdown

Den første regelen over sier at den deriverte av et konstant tall er null. Det gir forhåpentligvis mening ut fra definisjonen som sier at den deriverte er stigningstallet til tangenten. Stigningstallet til et konstant tall er null, og derfor er den deriverte null.

Disse fire reglene kommer vi til å bruke blant annet til å derivere polynomfunksjoner.

```{admonition} Eksempel 1: Derivasjon av polynom
:class: eksempel

Vi lar $f(x)=3x^4+2x^2-3$.

Her viser vi grundig hvilke derivasjonsregler vi bruker. Normalt sett vil vi her nesten bare skrive opp svaret uten mellomregninger.

$\begin{align}
f'(x)=(3x^4+2x^2-3)' &\stackrel{4}{=} (3x^4)'+(2x^2)'-(3)' \\
&\stackrel{3}{=} 3(x^4)'+2(x^2)'-(3)' \\
&\stackrel{1, 2}{=} 3\cdot 4x^{4-1}+2\cdot x^{2-1}-0 \\
&=12x^3+2x
\end{align}$
```

```{admonition} Eksempel 2: To viktige resultat
:class: eksempel
Resultatene under er vel verdt å huske både utledning av og resultatet. Det vil spare deg for mye tid framover.

$g(x)=\frac{1}{x}=x^{-1} \quad \Rightarrow \quad g'(x)=-x^{-1-1}=-x^{-2}=-\frac{1}{x^2}$

$h(x)=\sqrt{x}=x^{\frac{1}{2}} \quad \Rightarrow \quad h'(x)=\frac{1}{2}x^{\frac{1}{2}-1}=\frac{1}{2}x^{-\frac{1}{2}}=\frac{1}{2\sqrt{x}}$
```

Reglene over hjelper oss til å derivere alle potens-, rot- og polynomfunksjoner. Det er en stor gruppe av funksjoner, men vi har også arbeidet med logaritme- og eksponentialfunksjoner.

```{admonition} Derivert av logaritmer og eksponentialfunksjoner
:class: def

**Den naturlige logaritmen og den naturlige eksponentialfunksjonen**

$f(x)=\ln x \quad \Rightarrow \quad f'(x)=\frac{1}{x}$

$f(x)=e^x \quad \Rightarrow \quad f'(x)= e^x$

Beviset for $\ln x$ gir vi under. Bevis for $e^x$ kommer etter at vi har sett på kjerneregelen.

**Tierlogaritmen og generelle eksponentialfunksjoner**

$f(x)=\log x \quad \Rightarrow \quad f'(x)=\frac{1}{x\ln 10}$

$f(x)=a^x \quad \Rightarrow \quad f'(x)=a^x\cdot \ln a$

Den deriverte av tierlogaritmen gir vi under. Den deriverte av generelle eksponentialfunksjoner kommer vi tilbake til når vi har sett på kjerneregelen.
```

```{admonition} Bevis for den deriverte av $\ln x$
:class: def, dropdown
Vi ser på funksjonen $f(x)=\ln x$ og bruker definisjonen av den deriverte

$\begin{align}
f'(x)&=\lim_{\Delta x\to 0}\frac{f(x+\Delta x)-f(x)}{\Delta x} \\
&=\lim_{\Delta x\to 0}\frac{\ln(x+\Delta x)-\ln x}{\Delta x} \\
&=\lim_{\Delta x\to 0}\frac{\ln\left(\frac{x+\Delta x}{x}\right)}{\Delta x} \quad \text{andre logaritmesetning} \\
&=\lim_{\Delta x\to 0}\frac{1}{\Delta x}\ln\left(1\frac{\Delta x}{x}\right)
\end{align}$

Vi lar nå $n=\frac{x}{\Delta x}$ slik at $\frac{1}{\Delta x} =\frac{n}{x}$ og $\frac{\Delta x}{x}=\frac{1}{n}$. I tillegg vil grensen endres til $n\to\infty$.

$\begin{align}
f'(x)&=\lim_{n\to\infty}\frac{n}{x}\ln\left(1+\frac{1}{n}\right) \\
&=\lim_{n\to\infty}\frac{1}{x}\ln\left(1+\frac{1}{n}\right)^n \quad \text{tredje logaritmesetning} \\
&=\frac{1}{x}\lim_{n\to\infty}\left(1+\frac{1}{n}\right)^n \quad \text{trekker ut brøken uten n}\\
&=\frac{1}{x}\ln e \quad \text{grenseverdien var definisjonen av e} \\
&=\frac{1}{x}
\end{align}$

I siste linje har vi brukt at $\ln e=1$.
```

```{admonition} Bevis for den deriverte av $\log x$
:class: def, dropdown
Vi husker at $\log x=\frac{1}{\ln 10}\ln x$ og da kommer derivasjonsregelen over direkte.
```

Vi ser at den naturlige logaritmen har en forholdsvis enkel derivert som overraskende nok blir en potensfunksjon ($\frac{1}{x}=x^{-1}$).

Enda mer overraskende er at den deriverte av $e^x$ er $e^x$. Det viser seg at det er den eneste funksjonen vi kan lage (bortsett fra nullfunksjonen) som har en derivert som er lik seg selv. Og det er også begrunnelsen for at vi i det hele tatt valgte å innføre $e^x$ og $\ln x$. Ved å formulere eksponentialfunksjoner ved hjelp av $e$ vil derivasjon og analyse av vekstfart bli mye enklere. 

Stopp gjerne opp og tenk litt ekstra over hva resultatet $(e^x)'=e^x$ betyr. Vi husker at den deriverte i et punkt er lik stigningstallet til tangenten i punktet. Så dette betyr at tangenten til $e^x$ for $x=2$ har stigningstall $e^2$, tangenten for $x=5$ har stigningstall $e^5$ og tangenten for $x=\pi$ har stigningstall $e^{\pi}$. Med tanke på at $e$ er et tilsynelatende tilfeldig irrasjonalt tall er dette nærmest utrolig. Men det er også utrolig nyttig og tallet $e$ er derfor kanskje viktigere enn $\pi$ innenfor matematikken og naturvitenskapen.

```{admonition} Eksempel: Litt mer derivasjon
:class: eksempel

$f(x)=4e^x-2\ln x \quad \Rightarrow \quad f'(x)=4e^x-\frac{2}{x}$

$g(x)=\sqrt{x}-3e^x+\frac{3}{2}\ln x \quad \Rightarrow \quad g'(x)=\frac{1}{2\sqrt{x}}-3e^x+\frac{3}{2x}$
```

## Kjerneregelen
Med derivasjonsreglene vi har sett på til nå, så har vi grunnlaget for å derivere de aller fleste enkle funksjoner av typen:

 * $x^n$
 * $\sqrt{x}$ og andre røtter
 * $e^x$
 * $\ln x$

Vi kan gjøre mye med disse, men vi stopper opp med en gang vi møter uttrykk som $e^4x$, $\sqrt{x^3+2}$ eller $\ln(x^2+2)$. Dette er er eksempler på det vi kaller sammesatte funksjoner.

```{admonition} Eksempel 1: Sammensatt funksjon $e^{4x}$
:class: eksempel

$f(x)=e^{4x}$ er funksjonen $u(x)=4x$ satt inn i eksponentialfunksjonen $g(u)=e^u$. Vi kaller $u(x)$ den indre funksjonen (eller kjernen) og $g(u)$ den ytre funksjonen. Vi kan da skrive

$$f(x)=g(u(x))=e^{u(x)}=e^{4x}$$
```

```{admonition} Eksempel 2: Sammensatt funksjon $\sqrt{x^3+2}$
:class: eksempel

$f(x)=\sqrt{x^3+2}$ er funksjonen $u(x)=x^3+2$ satt inn i kvadratrotfunksjonen $g(u)=\sqrt{u}$. Vi kaller $u(x)$ den indre funksjonen (eller kjernen) og $g(u)$ den ytre funksjonen. Vi kan da skrive

$$f(x)=g(u(x))=\sqrt{u(x)}=\sqrt{x^3+2}$$
```

```{admonition} Eksempel 3: Sammensatt funksjon $\ln(x^2+2)$
:class: eksempel

$f(x)=\ln(x^2+2)$ er funksjonen $u(x)=x^2+2$ satt inn i logaritmefunksjonen $g(u)=\ln u$. Vi kaller $u(x)$ den indre funksjonen (eller kjernen) og $g(u)$ den ytre funksjonen. Vi kan da skrive

$$f(x)=g(u(x))=\ln u(x)=\ln(x^2+2)$$
```

Slike sammensatte funksjoner utgjør en stor gruppe av funksjoner og vi trenger en metode for å derivere de. Denne metoden kalles kjerneregelen.

```{admonition} Kjerneregelen
:class: def

La $f$ være en sammensatt funksjon med $g(u)$ som ytre funksjon og $u(x)$ som kjerne (indre funksjon). Den deriverte av $f$ er da:

$f(x)=g(u(x)) \quad \Rightarrow \quad f'(x)=g'(u(x))\cdot u'(x)$

Vi bruker kjerneregelen når den ytre funksjonen er en enkel funksjon som vi allerede kan å derivere med de grunnleggende reglene.
```

Kjerneregelen kan være vanskelig å forstå. Den sier at vi starter med å derivere den ytre funksjonen med hensyn på kjernen, før vi deriverer kjernen med hensyn på $x$. Forhåpentligvis vil noen eksempler gjøre det tydligere.

```{admonition} Eksempel 1: Kjerneregelen på $e^{4x}$
:class: eksempel
Vi ser på funksjonen $f(x)=e^{4x}$.
Kjernen er $u(x)=4x$, mens den ytre funksjonen er $g(u)=e^u$.

$f'(x)=g'(u(x))\cdot u'(x)=(e^u)'\cdot (4x)' =e^u\cdot 4=4e^{4x} $

Når vi kan kjerneregelen godt vil vi ofte gå gjøre dette direkte uten alle mellomregningene, men det anbefales å følge disse stegene fram til du har kontroll.
```

```{admonition} Eksempel 2: Kjerneregelen $\sqrt{x^3+2}$
:class: eksempel
Vi ser på funksjonen $f(x)=\sqrt{x^3+2}$.
Kjernen er $u(x)=x^3+2$, mens den ytre funksjonen er $g(u)=\sqrt(u)$.

$f'(x)=g'(u(x))\cdot u'(x)=(\sqrt{u})'\cdot (x^3+2)'=\frac{1}{2\sqrt{u}}\cdot 3x^2 =\frac{1}{2\sqrt{x^3+2}}\cdot 3x^2=\frac{3x^2}{2\sqrt{x^3+2}}$
```

```{admonition} Eksempel 3: Kjerneregelen på $\ln(x^2+2)$
:class: eksempel
Vi ser på funksjonen $f(x)=\ln(x^2+2)$.
Kjernen er $u(x)=x^2+2$, mens den ytre funksjonen er $g(u)=\ln u$.

$f'(x)=g'(u(x))\cdot u'(x)=(\ln u)'\cdot (x^2+2)' =\frac{1}{u}\cdot 2x = \frac{2x}{x^2+2}$
```

### Bevis for derivasjons av eksponentialfunksjoner

Vi kan bruke kjerneregelen til å bevise derivasjonsregelen $(e^x)'=e^x$.

```{admonition} Bevis for $(e^x)'=e^x$
:class: def
Vi setter $e^x=u$. Da får vi at $x = \ln u$. Vi deriverer begge sider og får

$\begin{align}
(x)'&=(\ln u)' \\
1&= \frac{1}{u}\cdot u' \\
u' &= u
\end{align}$

som gir at $(e^x)'=e^x$.
```

Videre beviser vi derivasjonsregelen for den generelle eksponentialfunksjon $a^x$.

```{admonition} Bevis for $(a^x)'=a^x\cdot\ln a$
:class: def
Vi lar $f(x)=a^x$. Videre utnytter vi at $a^x=e^{\ln a \cdot x}$. *Kontroller dette selv ved å bruke definisjonen av $\ln$*.

Vi lar $u(x)=\ln a \cdot x$ og $g(u)=e^u$. Det gir

$f'(x)=g'(u(x))\cdot u'(x)=e^u\cdot \ln a=e^{\ln a \cdot x}\cdot \ln a= a^x\cdot\ln a$
```

## Produktregelen
Kjerneregelen hjelper oss når en funksjon er satt inn i en annen funksjon. Vi kan også lage nye funksjoner ved å multiplisere sammen funksjoner. For å derivere slike funksjoner kan vi få bruk for produktregelen for derivasjon.

```{admonition} Produktregelen
:class: def
Vi lar $f(x)=u(x)\cdot v(x)$ være en funksjon som er produktet av to deriverbare funksjoner $u(x)$ og $v(x)$. Da vil den deriverte være gitt ved

$$f'(x)=u'(x)\cdot v(x)+u(x)\cdot v'(x)$$
```

Vi gjengir her kun starten og slutten på beviset. Mellom der er det mye algebra som de interesserte kan søke opp eller finne i de fleste lærebøker

```{admonition} Bevis for produktregelen
:class: def, dropdown
Vi tar utgangspunkt i en funksjon på formen $f(x)=u(x)\cdot v(x)$, der $u$ og $v$ er deriverbare funksjoner.

$\begin{align}
f'(x)&=\lim_{\Delta x\to 0}\frac{f(x+\Delta x)-f(x)}{\Delta x} \\
&=\lim_{\Delta x\to 0}\frac{u(x+\Delta x)\cdot v(x+\Delta x)-u(x)\cdot v(x)}{\Delta x} \\
\vdots \\
&=\lim_{\Delta x\to 0}\frac{u(x+\Delta x)-u(x)}{\Delta x}\cdot \lim_{\Delta x\to 0}\frac v(x\Delta x)+ \lim_{\Delta x\to 0}u(x)\lim_{\Delta x\to 0}\frac{v(x+\Delta x)-v(x)}{\Delta x} \\
&= u'(x)\cdot v(x)+u(x)\cdot v'(x)
\end{align}$

```

```{admonition} Eksempel 1: Produktregelen
:class: eksempel
Vi skal derivere $f(x)=x^2\cdot e^x$. Vi kaller $u(x)=x^2$ og $v(x)=e^x$.

$\begin{align}
f'(x)&=u'(x)\cdot v(x)+u(x)\cdot v'(x) \\
&= (x^2)'\cdot e^x+x^2\cdot(e^x)' \\
&= 2x\cdot e^x+x^2\cdot e^x \\
&= (x^2+2x)\cdot e^x
\end{align}$

Er vi sikker på framgangsmåte, så kan vi føre det her mer effektivt som

$f'(x)=2x\cdot e^x+x^2\cdot e^x = (x^2+2x)\cdot e^x$
```

```{admonition} Eksempel 2: Produktregelen
:class: eksempel
Vi skal derivere $f(x)=\sqrt{x}\cdot \ln x$. Vi kaller $u(x)=\sqrt{x}$ og $v(x)=\ln x$.

$\begin{align}
f'(x)&=u'(x)\cdot v(x)+u(x)\cdot v'(x) \\
&= (\sqrt{x})'\cdot \ln x+(\sqrt{x})\cdot(\ln x)' \\
&= \frac{1}{2\sqrt{x}}\cdot \ln x+\sqrt{x}\cdot \frac{1}{x} \\
&= \frac{\ln x}{2\sqrt{x}}+\frac{sqrt{x}}{x} \\
&= \frac{\ln x}{2\sqrt{x}}+\frac{1}{\sqrt{x}} \\
&= \frac{\ln x+2}{2\sqrt{x}}\\
\end{align}$

Er vi sikker på framgangsmåte, så kan vi føre det her mer effektivt som

$f'(x)=\frac{1}{2\sqrt{x}}\cdot \ln x+\sqrt{x}\cdot \frac{1}{x}=\frac{\ln x+2}{2\sqrt{x}}$
```

```{admonition} Eksempel 3: Produkt- og kjerneregelen
:class: eksempel
Vi skal derivere $f(x)=\sqrt{x^2+2}\cdot e^{x}$. Vi kaller $u(x)=\sqrt{x^2+2}$ og $v(x)=e^{x}$.

Vi trenger kjerneregelen for å derivere $u(x)$. Vi kaller kjernen for $w(x)=x^2+2$, mens den ytre funksjonen blir $g(w)=\sqrt{w}$. Da får vi

$u'(x)=g'(w(x))\cdot w'(x)=(\sqrt{w})'\cdot(x^2+2)' =\frac{1}{2\sqrt{w}}\cdot 2x =\frac{x}{\sqrt{x^2+2}}$

Så går vi videre med produktregelen på $f(x)$.

$\begin{align}
f'(x)&=u'(x)\cdot v(x)+u(x)\cdot v'(x) \\
&= (\sqrt{x^2+2})'\cdot e^x +(\sqrt{x^2+2})\cdot(e^x)' \\
&= \frac{x}{\sqrt{x^2+2}}\cdot e^x +\sqrt{x^2+2}\cdot e^x \\
&= \frac{e^x(x+1)}{\sqrt{x^2+2}}
\end{align}$

Mer kompakt føring vil ta kjerneregelen direkte uten mellomregning og hoppe over linje 1 og 2 i den siste utregningen.
```

## Brøkregelen
Den siste derivasjonsregelen vi skal ta for oss er brøkregelen. Med den har vi alle reglene vi skal lære og kan derivere stort sett alt som kan deriveres for hånd.

```{admonition} Brøkregelen
:class: def
Vi lar $f(x)=\frac{u(x)}{v(x)}$ være en funksjon gitt av kvotienten mellom to deriverbare funksjoner $u(x)$ og $v(x)$. Da vil den deriverte være gitt ved

$$f'(x)=\frac{u'(x)\cdot v(x)-u(x)\cdot v'(x)}{(v(x))^2}$$
```

Beviset for brøkregelen tar utgangspunkt i produktregelen (og kjerneregelen) og viser at brøkregelen strengt tatt er *unødvendig* da brøken i funksjonen kan skrives som et produkt $f(x)=u(x)\cdot \frac{1}{v(x)}$. Brøkregelen gir oss imidlertid en snarvei og vil forenkle en del problemer.

```{admonition} Bevis for brøkregelen
:class: def, dropdown
Vis skriver om $f(x)=\frac{u(x)}{v(x)}=u(x)\cdot \frac{1}{v(x)}$ og bruker produktregelen.

$\f'(x)=u'(x)\cdot \frac{1}{v(x)}+u(x)\cdot \frac{-1}{(v(x))^2}\cdot v'(x)=\frac{u'(x)\cdot v(x)-u(x)\cdot v'(x)}{(v(x))^2}$

Her har vi brukt kjerneregelen for å vise at 

$(\frac{1}{v(x)})'=(v(x)^{-1})'=\frac{-1}{(v(x))^2}\cdot v'(x)$
```

```{admonition} Eksempel 1: Brøkregelen
:class: eksempel
Vi skal derivere $f(x)=\frac{3x+2}{x^2}$.

$f'(x)=\frac{(3x+2)'\cdot x^2-(3x+2)\cdot (x^2)'}{(x^2)^2}=\frac{3x^2-(3x+2)\cdot 2x}{x^4}=\frac{-3x^2-4x}{x^4}=\frac{-3x-4}{x^3}$
```

```{admonition} Eksempel 2: Brøkregelen
:class: eksempel
Vi skal derivere $f(x)=\frac{e^{2x}}{x}$.

$f'(x)=\frac{(e^{2x})'\cdot x-e^{2x}\cdot x'}{x^2}=\frac{2e^{2x}\cdot x-e^{2x}\cdot 1}{x^2}=\frac{e^{2x}(x-1)}{x^2}$
```
