Vi har allerede sett litt på asymptoter, men vi skal her definere de grundigere ut fra grenseverdier.

```{adminition} Asymptoter
:class: def
En asymptote er en linje som grafen til en funksjon nærmer seg.

Vi kan ha:\
 * vertikale asymptoter
 * horisontale asymptoter
 * skrå asymptoter
```

## Vertikale asymptoter
Vertikale asymptoter kjenner vi allerede godt til fra rasjonale funskjoner, men vi skal her definere de skikkelig

```{admonition} Vertikal asymptote
:class: def
Linja $x=a$ er en vertikal asymptote for funksjonen $f(x)$ dersom en av betingelsene under stemmer.

$\displaystyle\lim_{x\to a^-} f(x) = \pm\infty$

$\displaystyle\lim_{x\to a^+} f(x) = \pm\infty$

For rasjonale funskjoner vil vi kunne ha vertikale asymptoter der nevneren har nullpunkt.
```

På forrige side så vi at $\displaystyle\lim_{x\to a^+} \ln (x-a)=-\infty$. Det vil si at den funksjonen har en vertikal asymptote for $x=a$. Ellers vil vi stort sett møte slike asynptoter hos rasjonale funksjoner. I 1T arbeidet vi stort sett med rasjonale funksjoner med en vertikal asymptote. Som vi ser i eksempelet under, så er det fullt mulig med mer enn en.

`````{admonition} Eksempel: To vertikale asymptoter
:class: eksempel
Vi skal finne de vertikale asymptotene til $f(x)=\frac{x}{x^2-9}$

Eventuelle vertikale asymptoter finner vi i nullpunktene til nevneren. Det vil si når $x=3\vee x=-3$. Vi sjekker med definisjonen av vertikal asymptote

$\displaystyle\lim_{x\to 3^+} \frac{x}{x^2-9} = \frac{3}{0}=\infty$

$\displaystyle\lim_{x\to -3^+} \frac{x}{x^2-9} = \frac{-3}{-0}=\infty$

Vi ser at begge grenseverdiene går mot uendelig.

Det vil si at det er vertikale asymptoter ved $x=3$ og ved $x=-3$. Under har vi tegnet inn grafen til funksjonen sammen med de vertikale asymptotene. Ved hjelp av kommandoen "Asymptote" kan GeoGebra tegne inn alle asymptoter for deg.

`````

```{admonition} Eksempel: Rasjonal funksjon uten vertikal asymptote
:class: eksempel
I forrige eksempel var vi nøye med å sjekke grenseverdiene før vi konkluderte med asymptotene. Grunnen til det er at vi kan ha nullpunkt i nevner som ikke gir vertikale asymptoter. Det skjer hvis vi samtidig har nullpunkt i teller. Da kan den problematiske faktoren forkortes vekk og vi får ingen vertikal asymptote. Vi så et eksempel på det i funksjonen $f(x)=\frac{2(x^2-9)}{x-3}$ på forrige side. Nevneren er null for $x=3$ og der kan det derfor være en vertikal asymptote. Ved å faktorisere og forkorte fant vi imidlertid ut at

$\displaystyle\lim_{x\to 3} f(x) = \frac{2(x-3)(x+3)}{x-3}=\lim_{x\to 3} 2(x+3)=\lim_{x\to 3} (2x+6)=2\cdot 3+6=12$

Siden grenseverdien eksisterer er det ingen vertikal asymptote for $x=3$.
```

Rasjonale funksjoner er såpass viktige at vi oppsummerer resultatene her.

```{admonition} Vertikale asymptoter til rasjonale funksjoner
:class: def
La funksjonen $f(x)$ være en rasjonal funksjon gitt ved $f(x)=\frac{t(x)}{n(x)}$ der $t(x)$ og $n(x)$ er polynomfunksjoner. 

Vertikale asymptoter finner vi ved å finne nullpunktene til nevneren etter at alle felles faktorer mellom teller og nevner er forkortet vekk.

Det kan derfor være et nullpunkt $x=a$ i opprinnelig nevner $n(x)$ som ikke gir vertikale asymptoter. Som vi har sett, skjer det når $t(x)$ og $n(x)$ har felles nullpunkt i $x=a$. Etter at faktoren $x-a$ er forkortet vekk, vil grenseverdien $\displaystyle\lim_{x\to a} f(x)$ eksistere og det vil ikke være vertikal asymptote for $x=a$.
```

## Horisontale asymptoter

```{admonition} Horisontal asymptote
:class: def

En horisontal asymptote er ei vannrett linje som en funksjon $f(x)$ nærmer seg når $x\to\pm\infty$. Matematisk skriver vi det som at $y=a$ er en horisontal asymptote for funksjonen $f(x)$ dersom

$\displaystyle\lim_{x\to\infty} f(x)=a$

og tilsvarende for $x\to -\infty$.
```

```{admonition} Eksempel: Horisontal asymptote
:class: eksempel

$\displaystyle\lim_{x\to\infty} (3^{-x}+2) = 0+2=2$

viser at funksjonen $f(x)=3^{-x}+2$ har en horisontal asymptote $y=2$.
```

Vi husker fra forrige side at rasjonale funksjoner har en grenseverdi når $x\to\infty$ kun dersom teller har lavere (da er grenseverdien 0) eller lik grad som nevner. Denne grenseverdien gir den horisontale asymptoten og for å finne den kan vi bruke trikset om å dele på høyeste potens av $x$ i alle ledd.

```{admonition} Eksempel: Horisontal asymptote for rasjonal funksjon
:class: eksempel

Vi vil finne den horisontale asymptoten til $f(x)=\frac{2x^2-3x+4}{3x^2-5}$.

$\begin{align}
\displaystyle\lim_{x\to\infty}\frac{2x^2-3x+4}{3x^2-5}&=\lim_{x\to\infty}\frac{2-\frac{3}{x}+\frac{4}{x^2}}{3-\frac{5}{x^2}}\\
&=\frac{2-0+0}{3-0} \\
&=\frac{2}{3}
\end{align}$

Den horisontale asymptoten er $y=\frac{2}{3}$.

```

I eksempelet over ville vi fått akkurat samme grenseverdi om vi hadde sett på $x\to -\infty$. Det kan imidlertid være lurt å sjekke begge grenseverdier som eksempelet under viser.

`````{admonition} Eksempel: To horisontale asymptoter
:class: eksempel
Vi skal finne asymptotene til $f(x)=\frac{2x}{x(e^{-x}+1)}$.

Det første vi observerer er at nevneren har et nullpunkt når $x=0$. Faktoren $e^{-x}+1$ er positiv for alle $x$ siden en eksponentialfunksjon aldri kan bli negativ. Nullpunktet i nevneren er imidlertid felles med telleren, så vi undersøker grenseverdien $x\to 0$ og får

$\displaystyle\lim_{x\to 0} \frac{2x}{x(e^{-x}+1)}=\lim_{x\to 0}\frac{2}{e^{-x}+1} = \frac{2}{e^0+1}=\frac{2}{1+1}=1$

Grenseverdien eksisterer, så vi har ingen vertikal asymptote.

Det neste vi undersøker er grenseverdien til $f(x)$ når $x\to\infty$.

$\displaystyle\lim_{x\to\infty}\frac{2x}{x(e^{-x}+1)}=\lim_{x\to\infty}\frac{2}{e^{-x}+1}=\frac{2}{0+1}=2$

Vi har altså en horisontal asymptote for $y=2$ når $x\to infty. Sjekker også $x\to -\infty$.

$\displaystyle\lim_{x\to -\infty}\frac{2x}{x(e^{-x}+1)}=\lim_{x\to -\infty}\frac{2}{e^{-x}+1}=\frac{2}{\infty+1}=0$

Det er altså en horisontal asymptote også for $y=0$ når $x\to -\infty$. Grafen til funksjonen ser slik ut.

```{figure} ./bilder/tohorisontale.png
---
scale: 20%
---
```
`````


## Skrå asymptoter
I tilfeller der teller har høyere grad enn nevner i en rasjonal funksjon, så vil vi ikke kunne få noen horisontal asymptote. Det er fordi teller vokser raskere enn nevner og funksjonen vil gå mot uendelig. I de tilfellene der graden til teller er bare en større enn graden til nevner vil vi imidlertid kunne ha noe som kalles for en skrå asymptote.

```{admonition} Skrå asymptote
:class: def
La $f(x)=\frac{t(x)}{n(x)}$ være en rasjonal funksjon. Dersom graden til $t$ er en høyere enn graden til $n$, så vil $f(x)$ ha en skrå asymptote $y=ax+b$. 

Formelt definerer vi den ved at $y=ax+b$ er en skrå asymptote til $f$ dersom

$\displaystyle\lim_{x\to\infty}(f(x)-(ax+b))=0$

I praksis finner vi den skrå asymptoten ved å utføre polynomdivisjon av $t(x)$ mot $n(x)$.
```

`````{admonition} Eksempel: Skrå asymptote
:class: eksempel
Vi skal finne asymptotene til $f(x)=\frac{x^3+2}{x^2-6x+9}$.

Vi starter med å se etter vertikale asymptoter. Nevneren er et fullstendig kvadrat $(x-3)^2$ med nullpunkt $x=3$. Telleren har ikke nullpunkt for $x=3$, så $x=3$ er en vertikal asymptote.

Videre ser vi at teller har grad en større enn nevner, så vi kan se etter skrå asymptote. For å finne den utfører vi polynomdivisjonen $x^3+2 : x^2-6x+9$. 

```{figure} ./bilder/polynomdivisjon.png
---
scale: 50%
---
```

Vi ser at polynomdivisjonen ikke går opp og vi får en rest. $f(x)$ kan nå skrives som $f(x)=x+6+\frac{27x-52}{x^2-6x+9}$. I grensen $x\to\infty$ vil brøken $\frac{27x-52}{x^2-6x+9}\to 0$. Vi kan da skrive

$\displaystyle\lim_{x\to\infty}\frac{x^3+2}{x^2-6x+9}=\lim_{x\to\infty}\left(x+6+\frac{27x-52}{x^2-6x+9}\right)=x+6+0=x+6\to\infty$

$f(x)$ går mot uendelig, men det går mot uendelig langs den rette linja gitt av $x+6$. Vi sier da at $y=x+6$ er en skrå asymptote for funksjonen.
`````
