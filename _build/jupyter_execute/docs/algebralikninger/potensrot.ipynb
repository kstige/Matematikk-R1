Vi jobbet også med en del regneregler for potenser.

```{admonition} Regneregler for potenser
:class: def

 1) $a^n \cdot a^m = a^{n+m}$
 2) $\frac{a^n}{a^m} = a^{n-m}$
 3) $\left(a\cdot b\right)^n = a^n \cdot a^m$
 4) $\left(\frac{a}{b}\right)^n = \frac{a^n}{b^n}$
 5) $\left(a^n\right)^m = a^{n\cdot m}$
 6) $a^0 = 1$
 7) $a^{-n} = \frac{1}{a^n}$

```
## Potenser med rasjonal eksponent

I definisjonen over ser vi at eksponenten må være et naturlig tall (positivt heltall) og med regneregel 6 og 7 kan vi også se på potenser med eksponent lik null eller et negativt tall. Vi kan oppsummere det med at $n\in\mathbf{Z}$.

Under har vi tegnet eksponentialfunksjonen $f(x)=2^x$. Hadde potenser kun vært definert for heltallige eksponenter ville grafen til funksjonen kun bestått av de blå punktene på grafen (heltallige x-verdier). Vi ser imidlertid at funksjonen også har verdier for alle x-verdiene mellom punktene. Det betyr at det må være mulig å definere potenser med andre eksponenter enn heltall. I R1 vil vi konsentrere oss om å definere potenser med rasjonale eksponenter (brøkuttrykk i eksponenten), men det kan også generaliseres til irrasjonale eksponenter.

```{figure} ./bilder/2x.png
---
scale: 40%
---
```

La oss starte med et eksempel for å se hva en rasjonal eksponent kan bety.

Vi ser på potensen $16^{\frac{1}{2}}$. I følge definisjonen vår skulle det betydd at 16 skal ganges med seg selv en halv gang (det betyr ikke at 16 skal ganges med en halv). Men hvordan kan et tall ganges med seg selv en halv gang? Det gir ikke mening. Vi må derfor angripe problemet på en annen måte og tar utgangspunkt i regnestykket under og regneregel 1 for potenser.

$\left(16^{\frac{1}{2}}\right)^2=16^{\frac{1}{2}}\cdot 16^{\frac{1}{2}}=16^{\frac{1}{2}+\frac{1}{2}}=16^1=16$

Vi ser altså at $16^{\frac{1}{2}}$ er det tallet som multiplisert med seg selv blir 16. Det tallet har et annet navn, nemlig kvadratroten av 16. Vi må derfor ha at $16^{\frac{1}{2}}=\sqrt{16}=4$. Generelt kan vi si at

```{admonition} Kvadratrot og potens
:class: def
Dersom $a\in\mathbf{R}$, så vil 

$a^{\frac{1}{2}}=\sqrt{a}$

Dette stemmer siden

$\left(a^{\frac{1}{2}}\right)^2=a^{\frac{1}{2}}\cdot a^{\frac{1}{2}}=a^{\frac{1}{2}+\frac{1}{2}}=a^1=a$
```

Resultatet vårt kan enkelt generaliseres til brøker av typen $\frac{1}{n}$ som eksponenter.

```{admonition} n-terøtter og potenser
:class: def
Dersom $a\in\mathbf{R}$ og $n\in \mathbf{Z}$, så vil 

$a^{\frac{1}{n}}=\sqrt[n]{a}$

Dette kalles n-teroten til a. n-teroten er det tallet som multiplisert med seg selv n ganger blir a.

$\left(\sqrt[n]{a}\right)^n=\underbrace{a^{\frac{1}{n}}\cdot a^{\frac{1}{n}} \cdots a^{\frac{1}{n}}}_{n \text{ ganger}}=a$
```

```{admonition} Eksempel: n-terot og potens
:class: eksempel

$\sqrt[3]{8}=2$ siden $2^3=8$. Vi kan da også skrive at $\sqrt[3]{8}=8^{\frac{1}{3}}=2$.

$\sqrt[4]{81}=3$ siden $3^4=81$. Vi kan da også skrive at $\sqrt[4]{81}=81^{\frac{1}{4}}=3$.
```

Alle brøkeksponentene våre så langt har hatt 1 i telleren. Det siste steget vi skal i generaliseringen av potenser er å se på eksponenter på formen $\frac{t}{n}$, altså et vilkårlig rasjonalt tall. Vi starter igjen med et eksempel for å lede oss på riktig vei.

Vi ser på potensen $16^{\frac{5}{2}}$ og bruker potensregel 5 baklengs. Da ser vi at

$16^{\frac{5}{2}}=16^{\frac{1}{2}\cdot 5} = \left(16^{\frac{5}{2}}\right)^5=\left(\sqrt{16}\right)^5$.

Alternativt kan vi gjøre det slik

$16^{\frac{5}{2}}=16^{5\cdot\frac{1}{2}}=\left(16^5\right)^{\frac{1}{2}}=\sqrt{16^5}$

Oppsummert har vi at

$16^{\frac{5}{2}}=\left(\sqrt{16}\right)^5=\sqrt{16^5}$

Generelt blir det

```{admonition} Generell definisjon av rasjonale eksponenter
:class: def
Dersom $a>0$ og $n, t \in \mathbf{Z}$, så vil 

$a^{\frac{t}{n}}=\left(\sqrt[n]{a}\right)^t= \sqrt[n]{a^t}$

Merk at definisjonen kun gjelder for grunntall større enn null.
```

Denne sammenhengen viser at potenser og n-terøtter egentlig er det samme rent matematisk. Alle regnereglene for potenser kan derfor omformuleres til tilsvarende regler for røtter. I de aller fleste tilfeller vil det imidlertid være enklest å heller omformulere problemet til potenser og bruke potensreglene. Vi oppgir derfor kun de viktigste regnereglene for røtter her.

```{admonition} Regneregler for n-terøtter
:class: def

$\sqrt[n]{ab}=\sqrt[n]{a}\sqrt[n]{b}$

$\sqrt[n]{\frac{a}{b}}=\frac{\sqrt[n]{a}}{\sqrt[n]{b}}$

Det er også verdt å merke seg at $\sqrt[n]{a+b}\neq\sqrt[n]{a}+\sqrt[n]{b}$.
```

Vi fortsetter med noen eksempler på bruk av sammenhengene vi har sett på.

```{admonition} Eksempler
:class: eksempel, dropdown

$\sqrt[4]{4^2}=\left(4^2\right)^{\frac{1}{4}}=4^{2\cdot \frac{1}{4}}=4^{\frac{1}{2}}=\sqrt{4}=2$

$a\sqrt[4]{a^3}\sqrt[4]{a^2}=a^1\cdot a^{\frac{3}{4}}\cdot a^{\frac{2}{4}}=a^{1+\frac{3}{4}+\frac{2}{4}}=a^{\frac{9}{4}}=\sqrt[4]{a^9}$

Svaret på denne kan også skrives som $a^{\frac{9}{4}}=a^{2+\frac{1}{4}}=a^2\cdot a^{\frac{1}{4}}=a^2\sqrt[4]{a}$. Det er ingen fasit på hva som er den enkleste/fineste måten å skrive svaret på. Bruk det som gjør det enklest å arbeide videre med uttrykket.

$
\frac{x\cdot \sqrt[3]{x}\cdot\sqrt[5]{x}}{\sqrt[15]{x^{13}}}=\frac{x^{1+\frac{1}{3}+\frac{1}{5}}}{x^{\frac{13}{15}}}=\frac{x^{\frac{23}{15}}}{x^{\frac{13}{15}}} =x^{\frac{23}{15}-\frac{13}{15}}=x^{\frac{10}{15}}=x^{\frac{2}{3}}=\sqrt[3]{x^2}$
```
