# L'Hôpitals regel

```{admonition} Læringsutbytte
:class: utbytte, dropdown
I dette temaet arbeider vi med kompetansemålet:
 * forstå begrepene vekstfart, grenseverdi, derivasjon og kontinuitet, og bruke disse for å løse praktiske problemer

Etter å ha arbeidet med temaet, skal du:
 * beregne grenseverdier ved å bruke L'Hôpitals regel
```

Stoffet på denne siden er litt på grensen til å være R1-pensum. Det kan tolkes inn under kompetansemålet over, men må ikke nødvendigvis bli det. 

L'Hôpitals regel er en metode (som bevises på universitet) for å beregne grenseverdier av $\frac{0}{0}$- og $\frac{\infty}{\infty}$-uttrykk.

```{admonition} L'Hôpitals regel
:class: def
Dersom grenseverdien

$\lim_{x\to a}\frac{f(x)}{g(x)}$

leder til et $\frac{0}{0}$- og $\frac{\infty}{\infty}$-uttrykk, så kan grenseverdien beregnes ved

$\lim_{x\to a}\frac{f(x){g(x)}}=\lim_{x\to a}\frac{f'(x)}{g'(x)}$
```

Vi tar noen eksempler:

````{admonition} Eksempel 1
:class: eksempel
Finn $\displaystyle{\lim_{x\to 3}\frac{2(x^2-9)}{x-3}}$.

```{admonition} Løsning
:class: ekslos, dropdown
Teller og nevner går mot null når $x$ går mot 3. På siden om grenseverdier beregnet vi denne grenseverdien ved å faktorisere teller og nevner og forkorte ved faktoren $x-3$. Nå kan vi gjøre det med L'Hôpitals regel.

$\begin{align}
\lim_{x\to 3}\frac{2(x^2-9)}{x-3}&=\lim_{x\to 3}\frac{(2(x^2-9))'}{(x-3)'} \\
&=\lim_{x\to 3}\frac{4x}{1} \\
&=\frac{4\cdot 3}{1} \\
&=12
\end{align}$
```
````

````{admonition} Eksempel 2
:class: eksempel
Finn $\displaystyle{\lim_{x\to 0}\frac{e^x-1}{2x}}$.

```{admonition} Løsning
:class: ekslos, dropdown
Teller og nevner går mot null når $x$ går mot 0. Bruker L'Hôpitals regel.

$\begin{align}
\lim_{x\to 0}\frac{e^x-1}{2x}&=\lim_{x\to 0}\frac{(e^x-1)'}{(2x)'} \\
&= \lim_{x\to 0}\frac{e^x}{2} \\
&= \lim_{x\to 0}\frac{e^0}{2} \\
&= \frac{1}{2}
\end{align}$
```
````

````{admonition} Eksempel 3
:class: eksempel
Finn $\displaystyle{\lim_{x\to \infty}\frac{e^x}{x^2}}$.

```{admonition} Løsning
:class: ekslos, dropdown
Teller og nevner går mot $\infty$ når $x$ går mot $\infty$. Bruker L'Hôpitals regel.

$\begin{align}
\lim_{x\to \infty}\frac{e^x}{x^2}&=\lim_{x\to \infty}\frac{(e^x)'}{(x^2)'} \\
&= \lim_{x\to \infty}\frac{e^x}{2x}  \quad \text{går mot uendelig, deriverer en gang til} \\
&= \lim_{x\to \infty}\frac{(e^x)'}{(2x)'} \\
&= \lim_{x\to \infty}\frac{e^x}{2} \\
&= \infty
\end{align}$

Grenseverdien går mot uendelig. Denne grenseverdien kunne vi ikke beregnet med vanlige metoder.
```
````

````{admonition} Eksempel 4
:class: eksempel
Finn $\displaystyle{\lim_{x\to \infty}e^{-x}\cdot \ln x^2}$.

```{admonition} Løsning
:class: ekslos, dropdown
Starter med å skrive om uttrykket

$\displaystyle\lim_{x\to \infty}e^{-x}\cdot \ln(x^2)=\lim_{x\to \infty}\frac{\ln x^2}{e^x}=\lim_{x\to \infty}\frac{2\ln x}{e^x}$

Teller og nevner går mot $\infty$ når $x$ går mot $\infty$. Bruker L'Hôpitals regel.

$\begin{align}
\lim_{x\to \infty}\frac{2\ln x}{e^x}&=\lim_{x\to \infty}\frac{(2\ln x)'}{(e^x)'} \\
&= \lim_{x\to \infty}\frac{\frac{1}{x}}{e^x} \\
&= \lim_{x\to 0}\frac{0}{\infty} \\
&= 0
\end{align}$
```
````
