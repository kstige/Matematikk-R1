# Omvendte funksjoner

```{admonition} Læringsutbytte
:class: utbytte, dropdown
I dette temaet arbeider vi med kompetansemålet:
 * utforske, analysere og derivere ulike funksjoner og deres omvendte funksjoner, og gjøre rede for egenskaper til og sammenhenger mellom slike funksjoner
 * analysere og tolke ulike funksjoner ved å bruke derivasjon

Etter å ha arbeidet med temaet, skal du:
 * forklare hva vi mener med omvendte (inverse) funskjoner
 * forklare sammenhengen mellom omvendte funksjoner
 * begrunne eksistens av omvendte funksjoner
 * kunne finne den omvendte funksjonen
 * derivere omvendte funksjoner
```

I matematikk har vi mange operasjoner som er omvendt (invers) av hverandre. For eksempel vil addisjon og subtraksjon, multiplikasjon og divisjon være omvendte operasjoner. Dersom vi tar utgangspunkt i en størrelse $x$, så vil $x +5-5$ og $\frac{x+5}{5} føre oss tilbake til $x$. Flere omvendte regneoperasjoner er:

 * $x^n$ og $\sqrt[n]{x}$ siden $\sqrt[n]{x^n}=x$
 * $10^x$ og $\lg x$ siden $10^{\lg x} = x$
 * $e^x$ og $\ln x$ siden $e^{\ln x} = x$

Her ser vi at en funksjon kan være den omvendte av en annen funksjon. Den ene funksjonen opphever effekten av den andre, eller fører oss tilbake til utgangspunktet. 

Vi kan for eksempel tenke oss en plante som er 5 cm da den plantes og vokser med 3 cm hver uke. En funksjon for høyden vil være $h(x)=3x +5$, der $x$ er antall uker. Den omvendte funksjonen til $h$ er funksjonen som gir oss antall uker når vi vet høyden. Vi skal snart vise at den funksjonen er $u(x)=\frac{x-5}{3}$ der $x$ er høyden i cm. En enkel test viser at:

$$h(4)=3\cdot 4+5=17$$

$$u(17)=\frac{17-5}{3}$$

Funksjonen $u(h)$ fører oss altså tilbake til antall uker. 

Før vi ser på flere eksempler definerer vi formelt hva vi mener med omvendte funksjoner.

```{admonition} Omvendte funksjoner
:class: def
To funksjoner $f$ og $g$ er omvendte funksjoner dersom

$f(g(x))=x$ for alle $x\in D_g$

$g(f(x))=x$ for alle $x\in D_f$

Skrivemåten $f(g(x))$ betyr at vi setter hele funksjonsuttrykket for $g(x)$ inn som $x$ i funksjonsuttrykket til $f$.

Det er også vanlig å kalle omvendte funksjoner for inverse funksjoner. Da vil man også se skrivemåten $f^{-1}$ for den omvendte funksjonen. Det er da viktig å være klar over at det ikke betyr det samme som $(f(x))^{-1}=\frac{1}{f(x)}$.
```

Vi skal nå bruke funksjonene $h$ og $u$ over som et eksempel på hvordan definisjonen fungerer.

````{admonition} Eksempel 1
:class: eksempel

Vis at funksjonene $h(x)=3x+5$ og $u(x)=\frac{x-5}{3}$ er omvendte funksjoner.

```{admonition} Løsning
:class: ekslos, dropdown
Vi ser at

$h(u(x))=3u+5=3\cdot \frac{x-5}{3} + 5 = x-5+5=x$

$u(h(x))=\frac{h-5}{3}=\frac{3x+5-5}{3}=\frac{3x}{3}=x$

Det betyr at $h$ og $u$ er omvendte funksjoner.
```
````

## Finne den omvendte funksjonen

I forrige eksempel viste vi at funksjonene $h(x)=3x+5$ og $u(x)=\frac{x-5}{3}$ er omvendte funksjoner, men hvordan kommer vi fram til uttrykket for den omvendte funksjonen?

La oss ta utgangspunkt i uttrykket for $h(x)$. For enkelthetsskyld setter vi $h(x)=y$. Så skriver vi om uttrykket slik at vi får $x$ uttrykt ved $y$.

$\begin{align}
y&=3x+5 \\
3x &= y-5 \\
x &= \frac{y-5}{3}
\end{align}$

Så bytter vi om på $x$ og $y$ slik at $u(x)=y=\frac{x-5}{3}$. Denne metoden kan vi bruke generelt så lenge funksjonen vi ser på har en omvendt funksjon. Det er det nemlig ikke alle funksjoner som har.

`````{admonition} Eksempel 2
:class: eksempel

**a)** Finn uttrykket $g(x)$ til den omvendte funksjonen til $f(x)=e^{2x+3}$.

**b)** Vis ved hjelp av definisjonen at $f$ og $g$ er omvendte funksjoner.

````{admonition} Løsning
:class: ekslos, dropdown

**a)** Vi setter $f(x)=y$ og prøver å finne $x$ uttrykt ved $y$.

$\begin{align}
y&=^{2x+3} \\
\ln y &= \ln e^{2x+3} \\
\ln y &= 2x+3 \\
x &= \frac{\ln y-3}{2}
\end{align}$

Så bytter vi om på $x$ og $y$ og får funksjonen $g(x)=\frac{\ln x -3}{2}$.

Vi kan også bruke CAS til å finne den omvendte funksjonen.

**b)** Bruker definisjonen får å sjekke at $f$ og $g$ er omvendte funksjoner.

$\displaystyle{f(g(x))=e^{2g+3}=e^{2\cdot(\frac{\ln x-3}{2})+3}=e^{\ln x -3 +3}=e^{\ln x}=x}$

$\displaystyle g(f(x))=\frac{\ln f -3}{2}=\frac{\ln \left(e^{2x+3}\right)-3}{2}=\frac{2x+3-3}{2}=\frac{2x}{2}=x$

Det viser at $f$ og $g$ er omvendte funksjoner.

```{admonition} Definisjonsmengde og verdimengde
:class: def

**Definisjonsmengden** er alle verdiene vi kan putte inn i funksjonen.\\
**Veridmengden** er alle verdiene vi kan få ut av funksjonen.

For omvendte funksjoner $f$ og $g$ har vi at

$D_f = V_g$ og $D_g = V_f$

`````

## Eksistens av omvendte funksjoner

I det siste eksempelet over så vi at vi måtte innskrenke definisjonsmengden til funksjonen $g(x)=x^2-2$ for at den skulle kunne være den omvendte funksjonen til $f(x)=\sqrt{x+2}$. Grunnen til det ser vi om vi prøver å speile hele $g(x)$ om linja $y=x$. Det kan vi gjøre i GeoGebra med verktøyet "Speil objekt om linje". Da får vi den røde kurva i grafen under

```{figure} ./bilder/speilinggraf.png
---
scale: 20%
---
```

Om vi husker tilbake til hvordan vi definerte funksjonsbegrepet i 1T, så ser vi at den røde kurva ikke er en funksjon. For en funksjon skal vi ha **kun** en verdi ut for hver verdi vi putter inn. Hvis vi putter inn $x=2$ vil vi her få ut både $2$ og $-2$. Det er derfor ikke en funksjon. Om vi innskrenker definisjonsmengden slik vi gjorde det i eksempelet oppstår ikke det problemet.

Ved å innskrenke definisjonsmengden gjør vi funksjonen *en-tydig*. Det er et krav for at den skal ha en ovendt funksjon.

```{admonition} Eksistens av omvendte funksjoner
:class: def
Funksjoner $f$ som er en-tydig har omvendte funksjoner.

At en funksjon $f$ er en-tydig betyr at det for hver $x$-verdi er kun en $y$-verdi og for hver $y$-verdi er kun en $x$-verdi. Eller sagt med symbol

$x_1 \neq x_2 \Rightarrow $f(x_1) \neq f(x_2)$ for alle $x_1, x_2 \in D_f$.

Dersom en funksjon skal være en-tydig kan den ikke ha topp- eller bunnpunkt. Det vil si at den deriverte ikke kan skifte fortegn. Det vil si at funksjonen enten må være strengt voksende eller strengt avtakende.
```

Litt mindre formelt kan vi si at hvis ingen horisontal linje krysser grafen til funksjonen i mer enn et punkt, så vil funksjonen ha en omvendt funksjon.

````{admonition} Eksempel 5
:class: eksempel
Undersøk om følgende funksjoner har en omvendt funksjon:

**a)** $f(x)=x^2$

**b)** $g(x)=e^x$

**c)** $h(x)=x^3+x^2+x$

```{admonition} Løsning
:class: ekslos, dropdown

**a)** Vi vet at $f(x)$ er en parabel som er symmetrisk om $y$-aksen. Det vil si at det er to $y$-verdier som gir samme $x$-verdi. Funksjonen er derfor ikke en-tydig og kan ikke ha en omvendt funksjon uten at vi begrenser definisjonsmengden. 

Om vi undersøker $f'(x)$, så ser vi at den skifter fortegn fra $-$ til $+$ for $x=0$. Dersom vi hadde *delt funksjonen i to* her, så kunne ville funksjonene med definisjonsmengde $\langle \leftarrow, 0]$ og $[0,\rightarrow\rangle$ hatt omvendte funksjoner.

**b)** Vi deriverer og får $g'(x)=e^x$. Siden eksponentialfunksjonen alltid er positiv vil $g'(x)>0$ og $g(x)$ er strengt voksende. Den har derfor en omvendt funksjon.

**c)** Vi deriverer og får $h'(x)=3x^2+2x+1$. Om vi prøver å løse $h'(x)=0$ enten med CAS eller abc-formelen, så får vi ingen løsning. Det vil si at $h'(x)$ aldri blir null og derfor heller ikke kan skifte fortegn. Da må $h(x)$ kunne ha en omvendt funksjon.
```
````


## Den deriverte av omvendte funksjoner
Til slutt vil vi ta for oss en sammenheng mellom den deriverte til to omvendte funksjoner. Vi beviser ikke sammenhengen, men vil antyde den med utgangspunkt i eksempelet vi startet med. Da så vi på de omvendte funksjonene $h(x)=3x+5$ og $u(x)=\frac{x-5}{3}$. Dette er to rette linjer og vi ser at $h'(x)=3$, mens $u'(x)=\frac{1}{3}$. Det kan se ut som at den deriverte av den ene er den inverse av den deriverte av den andre. Generelt har vi følgende sammenheng:

```{admonition} Derivasjon av omvendte funksjoner
:class: def
Dersom vi har to omvendte funksjoner $y=f(x)$ og $x=g(y)$, så vil sammenhengen mellom de deriverte funksjonene være:

$g'(y)=\frac{1}{f'(x)} \quad \text{når} \quad f'(x)\neq 0$

Definisjonen kan virke litt forvirrende og for noen vil den være lettere å forstå på følgende form:

$f'(x)=\frac{1}{g'(f(x))}$
```

Definisjonen kan være vanskelig å forstå i begge formene, men vi prøver med et eksempel. 

````{admonition} Eksempel 6
:class: eksempel
Vi har tidligere sett på de omvendte funksjonene $f(x)=e^{2x+3}$ og $g(x)=\frac{\ln x-3}{2}$. 

**a)** Vis at sammenhengen mellom de deriverte gjelder ved å derivere $f(x)$ direkte og ved hjelp av den omvendte funksjonen.

**a)** Vis at sammenhengen mellom de deriverte gjelder ved å derivere $g(x)$ direkte og ved hjelp av den omvendte funksjonen.

```{admonition} Løsning
:class: ekslos, dropdown

**a)** Direkte derivasjon ved hjelp av kjerneregelen:

$f'(x)=2e^{2x+3}$

Når vi skal derivere ved bruk av den omvendte funksjonen, så må vi først derivere $g(x)$:

$g'(x)=\frac{1}{2x}$

$f'(x)=\frac{1}{g'(f(x))}=\frac{1}{\frac{1}{2f(x)}}=\frac{1}{\frac{1}{2e^{2x+3}}}=2e^{2x+3}$

**b)** Har allerede at $g'(x)=\frac{1}{2x}$ og $f'(x)=2e^{2x+3}$. Bruker så derivasjon av omvendt funksjon for å finne $g'(x)$:

$g'(x)=\frac{1}{f'(g(x))}=\frac{1}{2e^{2g(x)+3}}=\frac{1}{2e^{2\cdot \frac{\ln x -3}{2}+3}}=\frac{1}{2e^{\ln x}}=\frac{1}{2x}$

I begge tilfeller ser vi at vi får det samme både ved direkte derivasjon og ved å bruke den omvendte funksjonen.
```
````

I eksempelet over ser vi at vi enklere kommer fram til svaret ved å bruke direkte derivasjon. Slik vil det ofte være. Fordelen med derivasjon ved hjelp av den omvendte er derfor i de tilfellene der det er vanskelig å finne et uttrykk for den omvendte funksjonen. Da kan vi bruke definisjonen over til å finne enkeltverdier av den derivert for den omvendte funksjonen.

````{admonition} Eksempel 7
:class: eksempel

Vi ser på funksjonen $f(x)=x^3+x^2+x$ og får vite at den går gjennom punktet $(1, 3)$. Finn $g'(3)$, der $g(x)$ er den omvendte funksjonen til $f$.

```{admonition} Løsning
:class: ekslos, dropdown

I eksempel 5 viste vi at funksjonen $f$ har en omvendt funksjon. Det er imidlertid vanskelig å finne et uttrykk for denne. Vi kan allikevel finne den deriverte i de oppgitte punktene. Finner først $f'(x)=3x^2+2x+1$. Viser videre hvordan begge former av definisjonen kan brukes til å finne $g'(3)$.

**Metode 1**

For den første metoden trenger vi at punktet $(1, 3)$ er på grafen til $f$. Da må $(3, 1)$ ligge på grafen til $g$.

$g'(3)=\frac{1}{f'(1)}=\frac{1}{3\cdot 1^2+2\cdot 1+1}=\frac{1}{6}$

**Metode 2**

$g'(3)=\frac{1}{f'(g(3))}=\frac{1}{f'(1)}=\frac{1}{3\cdot 1^2+2\cdot 1+1}=\frac{1}{6}$

Vi ser at eneste forskjellen egentlig er den første mellomregningen som "forklarer" bytte mellom $x$ og $y$.
```
````

````{admonition} Eksempel 8
:class: eksempel

**a)** Vi ser på funksjonen $f(x)=x^2-2$ med definisjonsmengden $D_f=[0, \rightarrow\rangle$. Finn $g'(5)$ når $g$ er den omvendte funksjonen til $f$.

**b)** Vi ser på samme funksjon, men med definisjonsmengden $D_f=\langle \leftarrow, 0]$. Finn $g'(5)$ når $g$ er den omvendte funksjonen til $f$.

```{admonition} Løsning
:class: ekslos, dropdown
**a)** Her bruker vi den første versjonen av definisjonen. Siden vi skal finne $g'(5)$ må vi ha $y=5$. Finner derfor først hvilken $x$-verdi det tilsvarer. Løser da $f(x)=5$ med abc-formelen og får $x=\pm 3$. Det er kun den positive verdien som ligger innenfor definisjonsmengden. Derfor må $y=5$ tilsvare $x=3$.

$g'(5)=\frac{1}{f'(3)}=\frac{1}{2\cdot 3}=\frac{1}{6}$

**b)** Her blir det helt tilsvarende, men $y=5$ må tilsvare $x=-3$.

$g'(5)=\frac{1}{f'(-3)}=\frac{1}{2\cdot (-3)}=-\frac{1}{6}$
```
````
