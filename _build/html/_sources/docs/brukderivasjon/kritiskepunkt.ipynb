## Stasjonære punkt

Vi husker fra 1T og den teoretiske gjennomgangen av den deriverte at verdien til den deriverte i et punkt gir oss stigningstallet til tangenten i punktet. Enklere forklart viser altså den deriverte oss hvor en funksjon stiger og hvor den synker. Dette kaller vi ofte for monotoniegenskapene til funksjonen.

```{admonition} Monotoniegenskaper
:class: def
Vi ser på en funksjon $f$. 

Dersom $f'(x)>0$ på et intervall $I$ sier vi at funksjonen er strengt voksende på intervallet. Da vil $f(b)>f(a)$ når $b>a$ og $a,b \in I$.

Dersom $f'(x)<0$ på et intervall $I$ sier vi at funksjonen er strengt avtakende på intervallet. Da vil $f(b)<f(a)$ når $b>a$ og $a,b \in I$.
```

De punktene der ingen av vilkårene over er oppfylt, så må enten $f'(x)=0$, eller $f'(x)$ er ikke definert. Vi tar for oss det første tilfellet først.

```{admonition} Stasjonære punkt
:class: def
Punkt på grafen der $f'(x)=0$ kaller vi stasjonære punkt. Vi har tre typer stasjonære punkt (b, t og s).

**Bunnpunkt:** $f'(b)=0$ og $f'(x)$ skifter fortegn fra negativ til positiv når vi passerer $b$.

**Toppunkt:** $f'(t)=0$ og $f'(x)$ skifter fortegn fra positiv til negativ når vi passerer $t$.

**Sadelpunkt/Terrassepunkt** $f'(s)=0$ og $f'(x)$ skifter ikke fortegn når vi passerer $s$.

De to første ($b$ og $t$) kalles ekstremalpunkt, mens funksjonsverdiene $f(b)$ og $f(t)$ kalles ekstremalverdier.

**Sett inn graf tilsvarenede s. 168 i boka.**
```

Vi tar et eksempel.

`````{admonition} Eksempel: Stasjonære punkt
:class: eksempel
Vi ser på funksjonen $f(x)=x^2 (x-2)^3$.
For å finne de stasjonære punktene må vi løse likningen $f'(x)=0$. Her er kanskje det enkleste å multiplisere ut parentesene slik at vi får et normalt enkelt polynom.

$\begin{align}
f(x)&=x^2 (x-2)^3=x^5-6x^4+12x^3-8x^2 \\
&\Rightarrow f'(x)=5x^4-24x^3+36x^2-16x
\end{align}$

Det er imidlertid ikke helt rett fram å løse likningen $f'(x)=0$ fra dette. Om vi er litt lure (og heldige) kan det bli enklere om vi deriverer det opprinnelige uttrykket med produkt- og kjerneregelen. Da holder vi det mest mulig faktorisert.

$\begin{align}
f'(x)&=2x\cdot(x-2)^3+x^2\cdot 3(x-2)^2\\
&=x(x-2)^2(2(x-2)+3x) \\
&=x(x-2)^2(5x-4) \\
&=5x(x-2)^2(x-\frac{4}{5})
\end{align}$

Vi kan da lese av løsningene til $f'(x)=0$ direkte og får de stasjonære punktene $x\in\{0, 2, \frac{4}{5}\}$.

For å avgjøre hva slags stasjonære punkt vi har kan vi for eksempel tegne ei fortegnslinje.


**Eksemepelet er ikke fullført. Mangler fortegnslinje.**
`````

`````{admonition} Eksempel: Stasjonært punkt med CAS
:class: eksempel
Vi ser på funksjonen $f(x)=x^2 (x-2)^3$ fra forrige eksempel og viser hvordan vi kan finne de stasjonære punktene med CAS.

Først definerer vi funksjonen, så løser vi likninga $f'(x)=0$.

```{figure} ./bilder/stasjonaerepunktCAS.png
---
scale: 50%
---
```

Vi får de samme stasjonære punktene som i forrige eksempel. For å avgjøre om det er topp- eller bunnpunkt, så kan vi finne verdien til den deriverte mellom de stasjonære punktene. Vi kan for eksempel se at $f'(x)$ er positiv i $x=-1$ som er en mindre $x$-verdi enn det første stasjonære punktet ($x=0$) og negativ i $x=\frac{1}{5}$ som ligger mellom $x=0$ og $x=\frac{4}{5}$. Det betyr at $f'(x)$ går fra positiv til negativ når vi passerer $x=0$, slik at $x=0$ må være et toppunkt. 

```{figure} ./bilder/fortegnCAS.png
---
scale: 50%
---
```

Tilsvarende kan vi finne $x=\frac{4}{5}$ er et bunnpunkt, mens $x=2$ er et sadelpunkt.

`````


## Kritiske punkt
De stasjonære punktene er en undergruppe av det vi kaller kritiske punkt.

```{admonition} Kritiske punkt
:class: def
Dersom $f$ er en funksjon med definisjonsmengde $D_f=[a,b]$, så vil de kritiske punktene til $f$ være:

 * alle punkter der $f'(x)=0$
 * alle punkter der $f$ ikke er deriverbar (bruddpunkt, knekkpunkt)
 * endepunktene $a$ og $b$

Om vi ser på et åpent intervall $\langle a, b \rangle$, så vil vi ikke ha endepunkt, og de vil derfor heller ikke være kritiske punkt.
```

```{admonition} Ekstremalpunkt
:class: def
Ekstremalpunktene (topp- og bunnpunktene) til en funksjon kan kun befinne seg i de kritiske punktene til funksjonen.

Merk: Det betyr ikke at et kritisk punkt må være et ekstremalpunkt.
```

Vi avslutter med et eksempel.

```{admonition} Eksempel: Kritiske punkt
:class: eksempel
Vi ser på funksjonen $f$ under og skal avgjøre hva ekstremalpunktene til funksjonen er.
$\begin{equation*}
    f(x) = \begin{cases} 
      x^2+1 & \text{når } \quad -3 \leq x < 2 \\
      -2x+9 & \text{når } x \geq 2
   \end{cases}
\end{equation*}$

Aller først, så ser vi at funksjonen har definisjonsmengde $D_f=[-3, \infty \rangle$. Det vil si at vi har et kritisk punkt i nedre grense $x=-3$.

Videre sjekker vi punktene der $f'(x)=0$. Vi deriverer hvert området for seg og ser at $f'(x)=0$ for $x=0$. Kontroller det selv.

I et tidligere eksempel på siden om kontinuitet har vi sett at denne funksjonen er kontinuerlig for $x=2$. Nå sjekker vi om funksjonen er deriverbar i punktet. Vi kontrollerer da om venstresidig og høyresidig grense for den deriverte er lik hverandre.

$\lim_{x\to 2^-}f'(x)=\lim_{x\to 2^-} 2x = 4$

$\lim_{x\to 2^+}f'(x)=\lim_{x\to 2^+}-2=-2$

Funksjonen er altså ikke deriverbar for $x=2$.

De kritiske punktene til funksjonen er altså $\{-3, 0, 2\}$. Det er da også kandidatene til ekstremalpunkt. For å avgjøre om det er topp- eller bunnpunkt tegner vi en fortegnslinje.


**Eksemepelet er ikke fullført. Mangler fortegnslinje.**
```
