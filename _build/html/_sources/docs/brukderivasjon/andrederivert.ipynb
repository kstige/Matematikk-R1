Vi har sett at når vi deriverer en funksjon $f(x)$, så ender vi opp med en ny funksjon $f'(x)$. Denne nye funksjonen viser hvor raskt $f(x)$ vokser. Men ellers er $f'(x)$ en vanlig funksjon på lik linje med $f(x)$. Det er derfor ingenting i veien for å derivere en gang til.

```{admonition} Den andrederiverte
:class: def
Den andrederiverte til en funksjon $f(x)$ er den deriverte av den deriverte funksjonen. Vi skriver det slik

$$(f'(x))'= f''(x)$$
```

Alt vi har lært til nå gjelder fortsatt. Vi bare deriverer en gang til. Det viktige for oss blir derfor hva den andrederiverte kan fortelle oss om den opprinnelige funksjonen.

## Krumning
$f''(x)$ viser oss altså om $f'(x)$ vokser eller avtar og hvor fort $f'(x)$ endrer seg. Dersom $f'(x)$ hadde vært konstant ($f''(x)=0$) ville stigningen aldri endre seg. Det gir ei rett linje. Om stigningen skal endre seg vil grafen til funksjonen krumme seg. Det vil si at verdien og fortegnet til $f''(x)$ må fortelle oss noe om hvordan $f(x)$ krummer seg.

```{admonition} Krumning
:class: def
Vi ser på en funksjon $f(x)$. 

Om $f''(x)>0$ vil $f'(x)$ øke og $f(x)$ være konveks (vende hul side opp).

Om $f''(x)<0$ vil $f'(x)$ avta og $f(x)$ være konkav (vende hul side ned).

```

Vi gir ikke noe bevis for påstandene over. Men om vi ser på bildet av grafen under og sammenlikner med fortegnslinje for $f(x)$, $f'(x)$ og $f''(x)$, så ser vi forhåpentligvis at det gir mening.


I punktet $P$ ser vi at krumningen til grafen går fra å vende nedover til å vende oppover. Et slikt punkt kaller vi et vendepunkt.

```{admonition} Vendepunkt og infleksjonspunkt
:class: def
Et vendepunkt til en graf er et punkt der grafen går fra å vende den hule siden ned til å vende den hule siden opp - eller motsatt. $x$-verdien til vendepunktet blir kalt for et *infleksjonspunkt* til funksjonen.

Vendepunkt (og infleksjonspunkt) finner vi der den andrederiverte skifter fortegn. For kontinuerlige funksjoner med kontinuerlige deriverte skjer det for de $x$-verdiene som er løsninger av $f''(x)=0$. Dersom det er punkt der funksjonen ikke er definert, bør vi også undersøke om den andrederiverte skifter fortegn i disse punktene.
```

## Vendepunktet
Siden den andrederiverte forteller oss om endringen i den deriverte, og vendepunktet forteller oss hvor denne endringen skifter fortegn, så må et av følgende skje i vendepunktet:

 * Grafen til funksjonen er brattest (øker eller avtar raskest)
 * Grafen til funksjonen er minst bratt (øker eller avtar saktest)

## Andrederiverttesten
Vi har sett hvordan vi kan bruke fortegnslinja til den deriverte for å avgjøre hvor en funksjon har topp- og bunnpunkt. Det er også mulig å benytte den andrederiverte til å avgjøre om et ekstremalpunkt er topp- eller bunnpunkt. For å gjøre det utnytter vi at grafen til en funksjon må ha hul side oppover ($f'' > 0$) når det er bunnpunkt og må ha hul side nedover ($f''<0$) når det er et toppunkt.

```{admonition} Andrederiverttesten
:class: def
Dersom $f'(x)=0$ og $f''(x) > 0$, så er $(x, f(x))$ et bunnpunkt.

Dersom $f'(x)=0$ og $f''(x) < 0$, så er $(x, f(x))$ et toppunkt.
```

``````{admonition} Eksempel: Andrederiverttesten
:class: eksempel
Finn ekstremalpunkt(ene) til $f(x)=x\cdot \ln(x)$. Bruk andrederiverttesten til å avgjøre om det er topp- eller bunnpunkt.

`````{admonition} Løsning
:class: ekslos, dropdown

Starter med å definere funksjonen i CAS. Finner så $x$-verdi til mulig ekstremalpunkt ved å løse $f'(x)=0$. Finner løsningen $x=\frac{1}{e}$. Finner at $f''(\frac{1}{e})=e>0$. Det betyr at punktet vi finner i linje 4 er et toppunkt.
`````` 
