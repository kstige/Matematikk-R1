# Den naturlige logaritmen

På forrige side så vi på tierlogaritmen. Den er spesielt viktig for oss siden vi har et titallssystem. I datamaskinverden er det imidlertid totallsystemet som gjelder. Da vil det være naturlig å bruke en logaritme med 2 som grunntall. Vi skriver da $lg_2$ og den fungerer helt likt som logaritmen med 10 som grunntall bare at den er definert ut fra en potens med 2 som grunntall. For eksempel vil $\lg_2 32=5$ siden $2^5=32$.

Slik kan vi definere logaritmer for alle grunntall vi ønsker, men vi kommer stort sett kun til å jobbe med tierlogaritmen og noe vi kaller den naturlige logaritmen. Den skal vi innføre snart, men først skal vi lære oss et nytt irrasjonalt tall som er minst like viktig som $\pi$.

## Den naturlige logaritmen
Siden tallet $e$ viser seg å være viktig og opptrer flere plasser i naturen, så har logaritmen med $e$ som grunntall fått navnet den naturlige logaritmen. Vi definerer den på helt samme måte som vi definerte tierlogaritmen.

```{admonition} Definisjon av den naturlige logaritmen
Den naturlige logaritmen, $\ln p$, til et tall $p$ er det tallet vi må opphøye $e$ i for å få $p$.

$e^{\ln p} = p$ der $p >0$
```

`````{admonition} Eksempel på logaritmeverdier
:class:

$\ln(e)=1$ siden $e^1=e$

$\ln(e^3) = 3$

$\ln(e^k) = k$

$\ln(\sqrt[3]{e^5})=\ln(e^{\frac{5}{3}})=\frac{5}{3}$

$\ln(5) \approx 1,61$ siden $e^{1,61}\approx 5$
`````
