# Tangenter

```{admonition} Læringsutbytte
:class: utbytte, dropdown
I dette temaet arbeider vi med kompetansemålet:
 * forstå begrepene vekstfart, grenseverdi, derivasjon og kontinuitet, og bruke disse for å løse praktiske problemer
 * analysere og tolke ulike funksjoner ved å bruke derivasjon

Etter å ha arbeidet med temaet, skal du:
 * finne tangenten til en funksjon i ulike punkt på grafen
```

Vi husker fra tidligere at tangenten til en funksjon i et punkt er den rette linja som akkurat berører funksjonen i punktet. Stigningstallet til tangenten er lik verdien til den deriverte i punktet. Her skal vi se nærmere på hvordan vi kan finne likninga til tangenten. Til det kan vi ha nytte av ettpunktsformelen som vi lærte om i 1T.

```{admonition} Ettpunktsformelen
:class: def

Dersom ei rett linje går gjennom punktet $(x_1, y_1)$ og har stigningstallet $a$, så vil likningen for den rette linja være gitt ved

$$y-y_1 = a(x-x_1)$$
```

Når vi skal finne tangenten til en funksjon for $x=x_1$ får vi at

 * stigningstallet $a=f'(x_1)$
 * tangeringspunktet $(x_1, y_1) = (x_1, f(x_1))$

Likningen for tangenten kan da finnes fra uttrykket

$$y-f(x_1)=f'(x_1)\cdot (x-x_1)$$

````{admonition} Eksempel 1
:class: eksempel

**a)** Finn likninga for tangenten til $f(x)=x^2+3x$ når $x=3$.

**b)** Finn skjæringspunktet til tangenten med $x$-aksen og $y$-aksen.

```{admonition} Løsning
:class: ekslos, dropdown
**a)**
Starter med å finne den deriverte i punktet og funksjonsverdien i punktet.

$f'(x)=2x+3 \Rightarrow f'(3)=2\cdot 3 + 3=9$

$f(3)=3^2+3\cdot 3=18$

Tangenten finner vi da ved ettpunktsformelen.

$\begin{align}
y-f(3)&=f'(3)(x-3) \\
y &= 9(x-3)+18 \\
y &= 9x-27+18 \\
y &= 9x-9
\end{align}$

**b)** 
Skjæringspunktet med $y$-aksen finner vi fra konstantleddet, så det er ved $y=-9$.

Skjæringspunktet med $x$-aksen finner vi ved å sette $9x-9=0$. Ser da at det er ved $x=1$.
```
````

`````{admonition} Eksempel 2

Vi ser på funksjonen $f(x)=x^3-3x^2+4$ fra eksempel 1 på forrige side. Finn den såkalte vendetangenten, altså tangenten i vendepunktet.

```{admonition} Løsning
:class: ekslos, dropdown
Vi har allerede funnet at vendepunktet er $(1, 2)$ og at den momentane vekstfarten i vendepunktet er -3 (se eksempel 1 på sida om funksjonsdrøfting). Ved bruk av ettpunktsformelen får vi da:

$\begin{align}
y - 2 &= -3(x-1) \\
y &= -3x +3+2\\
y &= -3x+5
\end{align}$
```
````{admonition} Løsning med CAS
:class: ekslos, dropdown
Vi definerer funksjonen i CAS. For polynomfunksjoner kan vi spare oss litt jobb med å bruke funksjonen *Vendepunkt* for å finne vendepunktet. For andre funksjoner må vi dobbeltderivere og løse likningen $f''(x)=0$ for å finne $x$-koordinat til vendepuntktet. Til slutt bruker vi kommandoen *Tangent* der vi setter inn $x$-verdien til vendepunktet og navnet til funksjonen. Får at vendetangenten er $y=-3x+5$
`````

`````{admonition} Eksempel 4
:class: eksempel
Vi ser på funksjonen $f(x)=4-x^2$ med definisjonsmengde $D_f=\langle 0, 2\rangle$.

Vi tegner inn en tangent til $f$ i punktet $P=(a, f(a))$. Tangenten skjærer $y$-aksen i $A$ og $x$-aksen i $B$. Sammen med origo $O$ danner $A$ og $B$ en rettvinklet trekant. Se figuren under.

```{figure} ./bilder/tangentgraf.png
---
scale: 50%
---
```

**a)** Finn likningen for tangenten.

**b)** Finn koordinatene til $A$ og $B$.

**c)** Finn et uttrykk for arealet til trekanten.

**d)** Hva er det minste arealet trekanten kan ha.

````{admonition} Løsning
:class: ekslos, dropdown

**a)** Tangenten finner vi ved å bruke ettpunktsformelen.

$\begin{align}
y-f(a)&=f'(a)(x-a) \\
y &= -2a(x-a)+(4-a^2) \\
y &= -2ax+a^2+4
\end{align}$

**b)** Skjæringspunkt med $y$-aksen når $x=0$ gir $A=(0, a^2+4)$.

Skjæringspunkt med $x$-aksen når $y=0$ gir likninga

$\begin{align}
0 &= -2ax+a^2+4 \\
2ax &= a^2+4 \\
x &= \frac{a^2+4}{2a}
\end{align}$

slik at $B=(\frac{a^2+4}{2a}, 0)$.

**c)** Arealet til trekanten er gitt ved 

$T = \frac{1}{2}\cdot (a^2+4)\cdot \frac{a^2+4}{2a} = \frac{(a^2+4)^2}{4a}$

**d)** For å finne det minste arealet trekanten kan ha, så må vi finne et bunnpunkt. Da deriverer vi uttrykket og setter det lik null. Her kan vi bruke brøkregelen for derivasjon, men uttrykket er så stygt at vi velger å bruke CAS (endrer da $a$ til $x$).

Definerer funksjonen i CAS, løser likninga $T'(x)=0$. Det er kun den positive løsninga som er innenfor definisjonsmengden. Den gir det kritiske punktet. Sjekker at den deriverte er negativ rett før og positiv rett etter det kritiske punktet. Det gir altså et bunnpunkt.

```{figure} ./bilder/tangentgrafCAS.png
---
scale: 50%
---
```

Det minste arealet trekanten kan ha er altså $\displaystyle\frac{32}{9}\sqrt{3}$.
`````
