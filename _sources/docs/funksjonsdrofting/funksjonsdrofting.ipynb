## Eksempler

Det vi har lært om på de to forrige sidene handler om å analysere eller drøfte funksjoner. Hva slags informasjon kan vi få ut av funksjonsuttrykket, og hva kan det sies om den situasjonen funksjonen beskriver. Her skal vi se på noen litt sammensatte eksempler der vi drøfter hva en funksjon sier oss.

`````{admonition} Eksempel 1
:class: eksempel
Vi ser på funksjonen $f(x)=x^3-3x^2+4$

**a)** Finn nullpunktene til funksjonen og avgjør hvor funksjonen er positiv/negativ.

**b)** Finn ekstremalpunktene og ekstremalverdiene til funksjonen.

**c)** Finn vendepunktet til funksjonen. Finn også verdien til funksjonen og den deriverte i dette punktet.

**d)** Skisser funksjonen.

````{admonition} Løsning
:class: ekslos, dropdown

**a)** Denne kan vi løse ved å prøve/feile oss fram til ei løsning, og så bruke polynomdivisjon for å faktorisere uttrykket. Her velger vi å bruke CAS.

```{figure} ./bilder/droftingeksempel1CAS.png
---
scale: 60%
---
```

Nullpunktene til funksjonen er $x=-1$ og $x=2$. Med det kan vi sette opp en faktorisering $f(x)=(x+1)(x-2)^2$. *Her må $x=2$ være et dobbelt nullpunkt siden vi ser i b) at det også er et bunnpunkt.* Tegner fortegnslinje for funksjonen. Den viser at funksjonen er positiv for $x\in\langle \leftarrow, -1\rangle \cup \langle 2, \rightarrow \rangle$ og negativ for $x\in\langle -1, 2\rangle$.

```{figure} ./bilder/droftingeksempel1f.png
---
scale: 50%
---
```

**b)** Vi starter med å derivere funksjonen før vi faktoriserer og leser av nullpunktene.

$f'(x)=3x^2-6x=3x(x-2)=0$

Ser da at den deriverte er $0$ for $x=0$ og $x=2$. Det er ingen andre kritiske punkt siden funksjonen er kontinuerlig og uten avgrensninger. For å avgjøre hva slags type stasjonært punkt det er, kan vi tegne fortegnslinje (eller bruke andrederiverttesten).

```{figure} ./bilder/droftingeksempel1df.png
---
scale: 50%
---
```

Det er altså:
 * Toppunkt i $x=0$ med funksjonsverdi $f(0)=4$
 * Bunnpunkt i $x=2$ med funksjonsverdi $f(2)=2^3-3\cdot 2^2+4=0$

 **c)** Vi starter med å derivere funksjonen igjen før vi fortsetter som i b).

 $f''(x)=6x-6=6(x-1)=0$

 Ser altså at den andrederiverte er $0$ for $x=1$. Sjekker med fortegnslinje om den andrederiverte også skifter fortegn i punktet.

```{figure} ./bilder/droftingeksempel1ddf.png
---
scale: 50%
---
```
$f''(x)$ skifter fortegn for $x=1$, så det må være et vendepunkt (strengt tatt infleksjonspunkt når vi kun snakker om $x$-verdien) der grafen skifter fra å være konkav til å bli konveks.

Regner ut funksjonsverdien og verdien til den deriverte i punktet.

$f(1)=1^3-3\cdot 1^2+4=2$ som gir at vendepunktet er i $(1,2)$.

$f'(1)=3\cdot 1^2 -6\cdot 1=-3$ som betyr at funksjonen har en momentan vekstfart på $-3$ i vendepunktet.

**d)** Tegner ei skisse av funksjonen basert på det vi har funnet ut i a-c).

```{figure} ./bilder/droftingeksempel1skisse.png
---
scale: 50%
---
```
Under skissen ser vi fortegnslinjene for $f(x)$, $f'(x)$ og $f''(x)$. Legg merke til hvordan informasjonen i linjene stemmer med fortegn, stigning og krumning på grafen til funksjonen.
````

`````

````{admonition} Eksempel 2
:class: eksempel
En funksjon er gitt ved $f(x)=x^3-3x^2+4x$.

*Merk: Funksjonen er bare nesten lik funksjonen i forrige eksempel.*

**a)** Når vokser funksjonen minst?

**b)** Hva er verdien til funksjonen da?

**c)** Hvor raskt vokser funksjonen da?

```{admonition} Løsning
:class: ekslos, dropdown

**a)** Deriverer først funksjonen to ganger.

$f'(x)=3x^2-6x+4$

$f''(x)=6x-6$

Setter så den andrederiverte lik null og finner at et mulig vendepunkt er for $x=1$. Ved å se at $f''(0)=-6$ og $f''(2)=6$, så ser vi også at $f''(x)$ skifter fortegn fra negativt til positivt i $x=1$. Det betyr at $f'(x)$ må ha et bunnpunkt der. Vi har altså minst vekst for $x=1$.

**b)** Verdien til funksjonen i vendepunktet er $f(1)=1^3-3\cdot 1^2+4\cdot 1=2$. Vendepunktet er altså i $(1,2)$.

**c)** Verdien til den deriverte i vendepunktet er $f'(1)=3\cdot 1^2-6\cdot 1+4=1$. Den momentane vekstfarten i vendepunktet er altså 1. Dette er den laveste verdien den deriverte har.
```
````

`````{admonition} Eksempel 3
:class: eksempel
Vi ser på funksjonen $f(x)=4\left(\lg x\right)^2-4$.

**a)** Finn nullpunktene til funksjonen.

**b)** Finn de kritiske punktene på grafen til funksjonen. Avgjør hva slags type punkt det er.

**c)** Finn infleksjonspunktet ($x$-verdien til vendepunktet) til funksjonen. Hva kan du si om funksjonen i vendepunktet?

**d)** Tegn grafene til $f(x)$, $f'(x)$ og $f''(x)$ i samme koordinatsystem og kommenter sammenhengene mellom grafene.

````{admonition} Løsning
:class: ekslos, dropdown

Før vi starter, så er det greit å merke seg at definisjonsmengden til $f(x)$ er $D_f = \langle 0, \rightarrow\rangle$ siden vi ikke kan ta logaritmen av negative tall eller null.

**a)** Løser likningen $f(x)=0$.

$\begin{align}
4\left(\lg x \right)^2-4&=0 \\
\left(\lg x\right)^2&=1 \\
\lg x &= \pm \sqrt{1}
\end{align}$

Vi får to muligheter videre

$\begin{align}
\lg(x)&=1 \quad &\vee \quad \lg x &= -1 \\
x &= 10 \quad &\vee x &= 0.1
\end{align}$ 

Det er altså nullpunkt for $x=0.1$ og $x=10$.

**b)** For å komme igang med den oppgava her minner vi om sammenhengen mellom $\lg$ og $\ln$ og hva den deriverte av $\ln x$ er.

$\lg x = \frac{\ln x}{\ln 10}$

$\left(\ln x\right)' = \frac{1}{x} \Rightarrow \left(\lg x\right)' = \frac{1}{(\ln 10) x}$

Deriverer $f(x)$ ved hjelp av sammenhengen over og kjerneregelen.

$f'(x)=4\cdot 2 \lg (x) \cdot (\lg x)' = 8\lg x \cdot \frac{1}{\ln 10 x} = \frac{8\ln x}{(\ln 10)^2 x}$

Løser videre $f'(x)=0$

$\begin{align}
\frac{8\ln x}{(\ln 10)^2 x}&=0 \\
\frac{\ln x}{x} &= 0 \\
\ln x & = 0 \quad x\neq 0 \\
x &= e^0 \\
x &= 1
\end{align}$

```{figure} ./bilder/droftingeksempel3df.png
---
scale: 50%
---
```

Det betyr at $f(x)$ har et bunnpunkt i det kritiske punktet $x=1$.

**c)** Deriverer en gang til. Må her bruke brøkregelen for derivasjon. Setter alle konstantene foran, slik at uttrykket i teller blir $\ln x$, mens uttrykket i nevner er $x$.

$f''(x)=\frac{8}{(\ln 10)^2}\cdot\frac{\frac{1}{x}\cdot x-\ln x \cdot 1}{x^2}=\frac{8}{(\ln 10)^2}\cdot\frac{1-\ln x}{x^2}$

Setter så $f''(x)=0$ og multipliserer begge sider med $x^2$. Det kan vi gjøre siden $x>0$. Får da at $\ln x= 1$ som gir at $x=e$. Tegner så fortegnslinje.

```{figure} ./bilder/droftingeksempel3ddf.png
---
scale: 50%
---
```

Siden $f''(x)$ skifter fortegn i $x=e$ må det være et infleksjonspunkt på grafen. Den deriverte må ha et toppunkt der siden $f''(x)$ går fra positiv til negativ. Det vil si at vendepunktet er der funksjonen vokser raskest.

**d)** Vi tegner grafen til $f(x)$ sammen med grafen til den deriverte og den andrederiverte i GeoGebra. Markerer inn

 * Nullpunktene $A$ og $B$ til $f(x)$
 * Bunnpunktet $C$ til $f(x)$ som samsvarer med nullpunktet $E$ til $f'(x)$
 * Vendepunktet $V$ til $f(x)$ som samsvarer med nullpunktet $F$ til $f''(x)$ og toppunktet $D$ til $f'(x)$

```{figure} ./bilder/droftingeksempel3graf.png
---
scale: 20%
---
```


````
`````
