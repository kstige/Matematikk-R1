# Logaritmelikninger

I ei logaritmelikning er den ukjente en del av et logaritmeuttrykk. Noen eksempler på logaritmelikninger er

 * $\lg x^2 = 9$
 * $2\ln(x-3)=4$
 * $(\ln x)^2-\ln x - 6 = 0$

Vi kommer til å få bruk for mange av de teknikkene som vi har lært om "vanlige" likninger tidligere. I tillegg kommer vi til å lære oss noen nye triks som baserer seg på blant annet logaritmesetningene.

En viktig ting vi må merke oss før vi fortsetter er at logaritmer kun er definert for positive tall. Det vil si at vi allerede fra start må sette en del begrensninger på hva den ukjente kan være. Eksempelvis vil vi i likningen $2\ln(x-3)=4$ måtte ha som et krav at $x>3$ siden en løsning som gir $x<3$ ikke kan brukes i den opprinnelige likningen.

For å lære oss å løse logaritmelikninger tar vi mange eksempel, der vi starter med det helt enkle og bygger oss opp til logaritmelikninger av høyere grad. Om logaritmelikningene inneholder $\lg$ eller $\ln$ har lite å si. Løsningsmetodene er helt lik, men i avsluttende steg må vi bruke riktig versjon av ekvivalensene som er oppgitt under. De er en versjon av definisjonen av logaritmene og blir nødvendig i løsningen av de fleste logaritmelikninger.

$$\lg x = a \quad \Leftrightarrow \quad x = 10^a$$

$$\ln x = a \quad \Leftrightarrow \quad x = e^a$$

```{admonition} Eksempel 1
:class: eksempel
I begge likningene under må $x>0$ siden vi tar henholdsvis $\lg x$ og $\ln x$. Om vi får et svar som gir $x<0$, så vil ikke svaret være gyldig.

$\begin{align}
\lg x &= 3 \\
x &= 10^3 \\
x &= 1000
\end{align}$

$\begin{align}
\ln x &= 3 \\
x &= e^3 
\end{align}$
```

```{admonition} Eksempel 2
:class: eksempel
Skal løse $3\ln x - 2 = 1$. Her må $x>0$.

$\begin{align}
3\ln x - 2 &= 1 \\
3\ln x &= 3 \\
\ln x &= 1 \\
x &= e^1 \\
x &= e
\end{align}$

```

```{admonition} Eksempel 2
:class: eksempel
Skal løse $2\lg x^2 - 2 = 2$. Her er eneste begrensning at $x\neq 0$ siden både positive og negative verdier for $x$ kan settes inn i likningen.

$\begin{align}
2\lg x^2 - 2 &= 2 \\
2\lg x^2 &= 4 \\
\lg x^2 &= 2 \\
x^2 &= 10^2 \\
x &= \pm \sqrt{10^2} \\
x &= \pm 10
\end{align}$

Det kan være fristende å velge en annen løsningsstrategi for likninga som tar utgangspunkt i tredje kvadratsetning.

$\begin{align}
2\lg x^2 - 2 &= 2 \\
2\cdot 2\lg x &= 4 \\
\lg x &= 1 \\
x &= 10^1 \\
x &= 10
\end{align}$

Her ser vi imidlertid at vi har mistet den negative løsningen. Grunnen til at dette skjer er at når vi tar logaritmen av $x^2$ kan vi fint ha $x<0$. Med en gang vi tar ned eksponenten ender vi imidlertid opp med å ta logaritmen av $x$. Da kan ikke $x$ være negativ lenger. Overgangen er derfor kun en implikasjon og ikke en ekvivalens. Problemet oppstår kun om eksponenten er et partall. Kan du tenke deg hvorfor?

```

```{admonition} Eksempel 3
:class: eksempel
Skal løse $\ln(2x-1)+\ln x = \ln 3$. Her ser vi at $2x-1>0$ og $x>0$. Den første gir $x>0,5$. Siden det er større enn 0 så blir det $x>0$ som gjelder.

For å løse denne bruker vi først andre logaritmesetning baklengs på venstre side.

$\begin{align}
\ln(2x-1)+\ln x &= \ln 3 \\
\ln((2x-1)\cdot x) &= \ln 3 \quad\quad \text{Tar e^ på begge sider} \\
e^{\ln((2x-1)\cdot x)} &= e^{\ln 3} \\
(2x-1)\cdot x &= 3 \\
2x^-x-3&=0
\end{align}$

Løser andregradslikningen med abc-formelen. Får at $x=-1 \vee x= frac{3}{2}$. Den negative løsningen kan ikke brukes i den opprinnelige likningen. Endelig løsning blir $L=\{\frac{3}{2}\}$.

I dette eksempelet ser vi at vi vi har fått innført en ekstra løsning som ikke stemmer. Det er fordi overgangen der vi fjerner logartimen kun er en implikasjon og ikke en ekvivalens. 

```

I linje 2-3 kunne vi tatt en snarvei ved å observere at: **Dersom logaritmen til to tall er like, så er tallene like**. Altså kunne vi skrevet opp $(2x-1)\cdot x=3$ direkte. Generelt har vi at

```{admonition} Like logaritmer
Dersom $a$ er et postivt tall, så vil

$\begin{align}
\lg x &= \lg a \\
x &= a
\end{align}$

$\begin{align}
\ln x &= \ln a \\
x &= a
\end{align}$
```

```{admonition} Eksempel 4
:class: eksempel
Vi skal løse $2\lg x^3 + 2\lg x^2 &= \lg 5$. Det kan vi gjøre på flere måter.

**Metode 1**

$\begin{align}
2\lg x^3 - 2\lg x^2 &= \lg 5 \\
6\lg x -4\lg x &= \lg 5 \\
2\lg x &=  \lg 5 \\
\lg x^2 &= \lg 5 \\
x^2 &= 5 \\
x &= \pm \sqrt{5}
\end{align}$

Her må vi imidlertid forkaste den negative løsningen siden $x^3<0$ når $x<0$. Da blir $L=\{\sqrt{5}\}$.

**Metode 2**

$\begin{align}
2\lg x^3 - 2\lg x^2 &= \lg 5 \\
\lg x^6 - \lg x^4 &= \lg 5 \\
\lg \frac{x^6}{x^4} &=  \lg 5 \\
\lg x^2 &= \lg 5 \\
x^2 &= 5 \\
x &= \pm \sqrt{5}
\end{align}$

Og vi får samme svaret som over $L=\{\sqrt{5}\}$.
```

```{admonition} Eksempel 5
:class: eksempel
Vi skal løse $\ln(2-x)-\ln x = 2$ med to forskjellige metoder. Her må $2-x>0$ og $x>0$. Det betyr at vi må ha $0<x<2$.

**Metode 1** - med andre logaritmesetning

$\begin{align}
\ln(2-x)-\ln x &= 2 \\
\ln \frac{2-x}{x} &= 2 \\
\frac{2-x}{x} &=  e^2 \\
2-x &= x\cdot e^2 \\
x(e^2+1) &= 2 \\
x &= \frac{2}{e^2+1}
\end{align}$

Finner vi et avrundet svar for $x$, så ser vi at det er en lovlig verdi, så $L=\{\frac{2}{e^2+1}\}$.\
Fra linje 3 til linje 4 multipliserte vi begge sider med $x$. Det er kun lov om $x\neq 0$, men det har vi uansett satt som en betingelse fra start, så det er ikke noe problem.

**Metode 2** - ta $e$ opphøyd i begge sider

$\begin{align}
\ln(2-x)-\ln x &= 2 \\
e^{\ln(2-x)-\ln x} &= e^2 \\
\frac{e^{\ln(2-x)}}{e^{\ln x}} &=  e^2 \\
\frac{2-x}{x} &=  e^2 
\end{align}$

Resten løses likt som metode 1 og vi får samme svar.
```

Til slutt tar vi for oss et eksempel på en likning som er i andregrad av logaritmeuttrykket

```{admonition} Eksempel: Andregradslikning
:class: eksempel
Vi skal løse $(\ln x)^2-2\ln x-8=0$. Her kan vi se for oss at dette er ei andregradslikning med $\ln x$ som ukjent. Dersom vi innfører $u = \ln x$ blir det kanskje enklere å se. Da blir likningen på formen $u^2-2u-8=0$. 

En slik likning løser vi greit med abc-formelen. Da får vi $u=2\vee u = -4$. Men så må vi sette inn igjen for $u$. 

$\begin{align}
u &=2 \quad &\vee& \quad u &=-4 \\
\ln x &= 2 \quad &\vee& \quad \ln x &= -4 \\
x &= e^2 \quad &\vee& \quad x &= e^{-4}
\end{align}$

Løsninga av likninga blir $L=\{e^{-4}, e^2\}$.
```
