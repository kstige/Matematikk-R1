# Eksponentiallikninger

```{admonition} Læringsutbytte
:class: utbytte, dropdown
I dette temaet arbeider vi med kompetansemålet:

 * utforske og forstå regneregler for potenser og logaritmer, og bruke ulike strategier for å løse eksponentialligninger og logaritmeligninger

Etter å ha arbeidet med temaet, skal du:
 
 * forklare hva en eksponentiallikning er
 * anvende ulike strategier for å løse eksponentiallikninger (blant annet ved hjelp av logaritmer)
 * forklare ekvivalens og implikasjon i logaritmiske likninger
 * bruke digitale hjelpemidler for å løse eksponentiallikninger
```

Eksponentiallikninger er likninger der den ukjente er i eksponenten i en potens. De enkleste eksponentiallikningene kan vi løse ved å gjøre høyre og venstre side om til en potens med samme grunntall. Da kan vi bruke følgende resultat som vi kjenner en tilsvarende form av fra logaritmelikninger.

```{admonition} Like grunntall
:class: def
Dersom $a^x=a^b$, så må $x=b$.
```

```{admonition} Eksempel 1
:class: eksempel
$\begin{align}
8^x &= 512 \\
8^x &= 8^3 \\
x &= 3
\end{align}$


$\begin{align}
e^{2x} &= \sqrt[3]{e^4} \\
e^{2x} &= e^{\frac{4}{3}} \\
2x &= \frac{4}{3} \\
x &= \frac{2}{3}
\end{align}$
```

For de fleste eskponentiallikninger vil vi imidlertid være avhengig av logaritmer for å løse likningene for hånd. Logaritmen bruker vi for å ta den ukjente ned fra eksponenten. Dette fungerer siden $\ln 4^x=x\ln 4$ i følge tredje logaritmesetning. Om ikke grunntallet i potensen tilsier at det er lurt å velge enten tierlogaritmen eller den naturlige logaritmen, så kan vi i utgangspunktet velge akkurat den logaritmen vi vil.

```{admonition} Eksempel 2
:class: eksempel

$\begin{align}
3^x &= 8 \\
\lg 3^x &= \lg 8 \\
x\lg 3 &= \lg 8 \\
x &= \frac{\lg 8}{\lg 3}
\end{align}$

Oppgava kan også løses ved å bruke $\ln$.

$\begin{align}
3^x &= 8 \\
\ln 3^x &= \ln 8 \\
x\ln 3 &= \ln 8 \\
x &= \frac{\ln 8}{\ln 3}
\end{align}$

Vi ser at svaret ser ulikt ut, men om vi finner tilnærmet verdi for de to uttrykkene, så ser vi at $\frac{\lg 8}{\lg 3}=\frac{\ln 8}{\ln 3}$. Det her kan vi også vise ved å ta i bruk sammenhengen mellom logaritmene som vi viste på siden om den naturlige logaritmen.
```

```{admonition} Eksempel 3
:class: eksempel

$\begin{align}
4^{2x+1} &= e^2 \\
\ln 4^{2x+1} &= \ln e^2 \\
(2x+1)\ln 4 &= 2 \\
2x+1 &= \frac{2}{\ln 4} \\
2x &= \frac{2}{\ln 4} -1 \\
x &= \frac{1}{\ln 4} -\frac{1}{2} \\
x &= \frac{2-\ln 4}{2\ln{4}} = \frac{2-\ln 4}{\ln 16}
\end{align}$
```

I enkelte eksponentiallikninger må vi gjøre litt forarbeid for å få likningen på en form som gjør at vi kan få eksponenten ned ved hjelp av en logaritme. Dette forarbeidet kan ligge i å slå sammen ulike potenser for å få en potens med den ukjente i, faktorisere eller få uttrykket på form av en andregradslikning. Ofte kreves det litt kreativitet og/eller erfaring for å se hvordan vi kan skrive om likningen for å gjøre det lettere for oss. Vi tar noen eksempler på ulike triks vi kan bruke.

```{admonition} Eksempel 4
:class: eksempel
Vi skal løse likningen $e^{2x} -4e^x = 0$. Et triks vi kan bruke her er at $e^{2x}=(e^x)^2=e^x\cdot e^x$. Dette bruker vi for å faktorisere uttrykket på venstre side.

$\begin{align}
e^{2x} -4e^x &= 0 \\
e^x\cdot e^x-4e^x &= 0 \\
e^x(e^x-4) &= 0
\end{align}$

Her kan vi videre benytte produktregelen vi lærte i 1T. For at $e^x(e^x-4)$ skal bli null, så må en av faktorene være null. Den første faktoren $e^x$ kan ikke bli null, så da går vi videre med

$\begin{align}
e^x -4 &= 0 \\
e^x &= 4 \\
\ln e^x &= \ln 4 \\
x &= \ln 4
\end{align}$
```

```{admonition} Eksempel 5
:class: eksempel

$\begin{align}
3^{3x}+3^{3x}+3^{3x} &= 27 \\
3\cdot 3^{3x} &= 27 \\
3^{3x} &= 9 \\
3^{3x} &= 3^2 \\
3x &= 2 \\
x &= \frac{2}{3}
\end{align}$
```

```{admonition} Eksempel 6
:class: eksempel 
Vi skal løse $3^x-4=12\cdot 3^{-x}$. Her vil vi først kvitte oss med potensen med negativ eksponent. Det gjør vi ved å multiplisere begge sider av likningen med $3^x$. Da får vi $(3^x)^2-4\cdot 3^x=12$. Likningen virker fortsatt vanskelig, men om vi foreløpig setter $u=3^x$ kan vi skrive den om til $u^2-4u = 12$. Det er ei standard andregradslikning som vi løser ved å flytte alt til venstre side og så bruke abc-formelen. Den gir $u=-2 \vee u=6$. Så setter vi inn for $u$ og går videre med

$\begin{align}
u &=-2 \quad &\vee& \quad u &=6 \\
3^x &= -2 \quad &\vee& \quad 3^x &= 6 \\
\text{går ikke} & \quad &\vee& \quad x &= \frac{\ln 6}{\ln 3}
\end{align}$

Det betyr at $L=\{\frac{\ln 6}{\ln 3}\}$.
```

Alle likningene her kan selvfølgelig også løses med CAS eller grafisk. Ved løsning med CAS vil vi kunne få svar som ser litt rare ut. Grunnen er at GeoGebra ønsker å skrive alle logaritmer uttrykt med $\ln$.
