## Tierlogaritmen

I 1T lærte vi mye om ulike funksjonstyper. En av funksjonstypene var eksponentialfunksjoner. Det er funksjoner på formen $f(x)=a\cdot b^x$, der $a,b\in\mathbf{R} \backslash \{0\}$. Vi husker at funksjonen representerte en størrelse med startverdi $a$ som økte med en fast prosentstørrelse når $x$ økte med 1. Denne prosenten kommer inn gjennom tallet $b$ som er en vekstfaktor.

I grafen under er eksponentialfunksjonen $f(x)=10^x$ tegnet inn.

```{figure} ./bilder/10x.png
---
scale: 30%

```

Fra grafen kan vi lese av ting som at:

 * $10^{-0,2}\approx 0.6$
 * $10^0 = 1$ 
 * $10^{0,3} \approx 2$

Vi kan også lese av den andre veien for å finne at:

 * For å få ut 0,5 må vi opphøye 10 i omtrent 0,3
 * For å få ut 4 må vi opphøye 10 i omtrent 0.6

Det vil si at vi kan bruke grafen til å løse likninger av typen $10^x=a$, der $a\in\mathbf{R}$. Om vi skal løse likninga $10^x=5$ kan vi for eksempel lage oss ei linje ut fra y=5. Den skjærer grafen for $x\approx 0,7$ som blir løsninga på likninga.

Tallsystemet vi bruker kalles titallsystemet. Potenser med 10 som grunntall dukker derfor ofte opp. Likninger av typen over er derfor en type likninger som vil være relvant å se på. $x$-verdien i likninga har derfor fått et eget navn som vi kaller den tierlogartimen (eller den briggske logaritmen). Vi starter med å definere dette begrepet.

```{admonition} Definisjon av tierlogaritmen
:class: def

Tierlogaritmen, $\lg p$, til et tall $p$ er det tallet vi må opphøye 10 i for å få $p$.

$10^{\lg p} = p$ der $p >0$

Skrivemåten *log* brukes av mange og betyr det samme som *lg*.
```

Tierlogaritmen kalles ofte for den **briggske logaritmen** etter matematikeren Henry Briggs som i 1617 ga ut en tabell med tierlogaritmen for de 1000 første naturlige tallene med en nøyaktighet på 14 desimaler. I dag bruker vi digitale hjelpemidler for å finne de aller fleste logaritmeverdier.

`````{admonition} Eksempel på logaritmeverdier
:class: eksempel

$\lg(100)=2$ siden $10^2=100$

$\lg(10000) = 4$ siden $10^4=10000$

$\lg(0,01)=-2$ siden $10^{-2}=0,01$

$\lg(5) \approx 0,69897$ siden $10^{0,69897}\approx 5$

De tre første kan vi finne uten hjelpemidler, men for de fleste (som den siste) kan vi bruke digitale hjelpemidler. Her er det vist med CAS og programmering i Python
`````

## Bruk av logaritmer
Logaritmer har færre bruksområder i dag enn det har hatt før. Regnereglene vi har sett på viser at logaritmer kan gjøre multiplikasjon om til addisjon og divisjon om til subtraksjon. Dette har historisk vært til stor hjelp før kalkulatorer og datamaskiner kunne utføre beregningene.

Logartimer er nødvendig for å løse liknigner der den ukjente befinner seg i eksponenten i en potens. Slike likninger kalles eksponentiallikninger og vi vil komme tilbake til de senere.

Logaritmer brukes til å definere en del størrelser i kjemi og fysikk:
 * pH er en skala i kjemi som forteller om et stoff er surt eller basisk. pH-skalaen går fra 1 til 14. En blanding med pH på 1 er 10 ganger så sur som en blanding med pH2. Dette defineres med logaritmer.
 * Richters skala for jordskjelv fungerer tilsvarende
 * dB-skalaen for lydstyrke er logaritmisk
