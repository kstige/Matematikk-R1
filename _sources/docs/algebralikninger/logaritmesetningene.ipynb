## Første logaritmesetning

```{admonition} Første logaritmesetning
:class: def

La $a$ og $b$ være positive tall. Da er

$\lg ab = \lg a + \lg b$

Logaritmer gjør altså multiplikasjon om til addisjon.

*Det er viktig å merke seg denne setningen. En vanlig feil er å si at $\lg(a+b)=\lg a + \lg b$.*
```

```{admonition} Bevis for første logaritmesetning
:class: def, dropdown
Utgangspunktet for beviset er definisjonen av logaritmer og regnereglene for potenser.

$ab=10^{\lg ab}$

$ab=a\cdot b = 10^{\lg a}\cdot 10^{\lg b}=10^{\lg a + \lg b}$

Begge resultatene våre er lik $ab$. De må derfor også være lik hverandre.

$10^{\lg ab}=10^{\lg a + \lg b}$

For at to potenser med samme grunntall skal være like, så må eksponentene være like. Altså vil

$\lg ab = \lg a + \lg b$
```

````{admonition} Eksempel 1: Første logaritmesetning
:class: eksempel
Forenkle uttrykket så mye som mulig

$\lg (1000x) + \lg (10x)$

```{admonition} Løsning
:class: ekslos, dropdown
$
\begin{align}
\lg (1000x) + \lg (10x) &= \lg (10^3\cdot x) + \lg (10\cdot x) \\
&=\lg 10^3+\lg x + \lg 10 + \lg x \\
&= 3+\lg x + 1 + \lg x \\
& = 4 +2\lg x
\end{align}$
```
````


````{admonition} Eksempel 2: Første logaritmesetning
:class: eksempel
Du får vite at $\lg 30 = 1,48$ og at $\lg 6 = 0,78$. Finn $\lg 180$.

```{admonition} Løsning
:class: ekslos, dropdown

$\lg 180 = \lg (6\cdot 30) = \lg 6 + \lg 30 = 0,78+1,48=2,26$
```
````

Andre logaritmesetning er ganske tilsvarende den første, men gjør om divisjon til subtraksjon.


## Andre logaritmesetning
```{admonition} Andre logartimesetning
:class: def
La $a$ og $b$ være positive tall. Da er

$\lg \frac{a}{b} = \lg a - \lg b$
```

Beviset er helt tilsvarende som beviset for første logaritmesetning og overlates til leseren.

````{admonition} Eksempel: Andre logaritmesetning
:class: eksempel
Forenkle uttrykket så mye som mulig

$\lg \frac{1000}{x} + 3\lg \frac{x}{100}$

```{admonition} Løsning
:class: ekslos, dropdown
$
\begin{align}
\lg \frac{1000}{x} + 3\lg \frac{x}{100} &= \lg \frac{10^3}{x} + 3\lg \frac{x}{10^2} \\
&=\lg 10^3-\lg x +3(\lg x-\lg 10^2) \\
&= 3-\lg x+3\lg x - 3\cdot 2 \\
& = 2\lg x -3
\end{align}$
```
````

## Tredje logaritmesetning
```{admonition} Tredje logaritmesetning
:class: def

La $a$ være et positivt tall og $b\in\mathbf{R}$. Da er

$\lg a^b = b\cdot \lg a$
```

En vanlig feil ved bruk av tredje logaritmesetning er å ikke skille mellom uttrykkene $\lg a^2=2\lg a$ og $\left(lg a\right)^2=\lg a \cdot \lg a$. Tredje logaritmesetning kan brukes på den første, men ikke den andre.

```{admonition} Bevis for tredje logaritmesetning
:class: def, dropdown
Vi tar utgangspunkt i definisjonen av logaritmer og av potensreglene.

$a^b=10^{\lg a^b}$
$a^b= \left(10^{\lg a}\right)^b=10^{b\cdot \lg a}$

Uttrykkene må være like hverandre slik at $10^{\lg a^b}=10^{b\cdot \lg a}$.

For at to potenser med samme grunntall skal være like, så må eksponentene være like. Altså vil

$\lg a^b = b\cdot \lg a$
```

````{admonition} Eksempel: Tredje logaritmesetning
:class: eksempel
Forenkle uttrykket så mye som mulig

$2\lg x^4 +3\lg \sqrt[3]{x^4}$

```{admonition} Løsning
:class: ekslos, dropdown
$
\begin{align}
2\lg x^4 +3\lg \sqrt[3]{x^4} &= 2\lg x^4+3\lg x^{\frac{4}{3}} \\
&=2\cdot 4\lg x+ 3\cdot \frac{4}{3}\lg{x} \\
&= 8\lg x +4\lg x \\
&= 12\lg x
\end{align}$
```
```` 

Vi avslutter med et sammensatt eksempel på bruk av logaritmesetningene.

````{admonition} Eksempel: Sammensatt bruk av logaritmesetningene
:class: eksempel
Forenkle uttrykket så mye som mulig

$3\lg(x^2\cdot y)-\frac{1}{3}\lg \left(\frac{(10\cdot y)^9}{x^3}\right)$

```{admonition} Løsning
:class: ekslos, dropdown
$\begin{align}
3\lg(x^2\cdot y)-\frac{1}{3}\lg \left(\frac{(10\cdot y)^9}{x^3}\right) &= 3(\lg x^2 + \lg y) -\frac{1}{3}(\lg 10^9 +\lg y^9 -\lg x^3) \\
&= 3\cdot 2\lg x +3\lg y -\frac{1}{3}(9+9\lg y -3\lg x) \\
&= 6\lg x +3\lg y -3 -3\lg y +\lg x \\
&= 7\lg x -3
\end{align}
$
```
````

## Den naturlige logaritmen
Logaritmesetningene som vi beviste for tierlogaritmen gjelder også for den naturlige logaritmen. Bevisene er også helt tilsvarende, så her oppgir vi kun resultat og noen eksempler.

```{admonition} Logaritmesetningene
:class: def
**Første logaritmesetning**\
For $a, b > 0$

$\ln ab = \ln a + \ln b$

**Andre logaritmesetning**\
For $a, b > 0$

$\ln \frac{a}{b}= \ln a - \ln b$

**Tredje logaritmesetning**\
For $a>0$ og $b\in \mathbf{R}$

$\ln a^b = b\ln a$
```

````{admonition} Eksempel 1
:class: eksempel
Forenkle uttrykket så mye som mulig

$\ln (5 e^2)-3\ln \frac{5}{e^3}$

```{admonition} Løsning
:class: ekslos, dropdown
$\begin{align}
\ln (5 e^2)-3\ln \frac{5}{e^3} &= \ln 5 + \ln e^2 -3(\ln 5-\ln e^3) \\
&= \ln 5 + 2 -3\ln 5 + 3\cdot 3 \\
&= 11 - 2\ln 5
\end{align}$
```
````

````{admonition} Eksempel 2
:class: eksempel
Forenkle uttrykket så mye som mulig

$\ln (a^{-2}\cdot e^4)-2\ln \frac{a}{a^2\cdot e^{-2}}$

```{admonition} Løsning
:class: ekslos, dropdown
$\begin{align}
\ln (a^{-2}\cdot e^4)-2\ln \frac{a}{a^2\cdot e^{-2}} &= \ln a^{-2} + \ln e^4 - 2(\ln a-\ln a^2 -\ln e^{-2}) \\
&= -2\ln a + 4 - 2\ln a -2\cdot (-2)\ln a -2\cdot(-1)\cdot(-2) \\
&= 0
\end{align}$
```
```` 

## Sammenheng mellom logaritmer
Det er ikke ofte vi får bruk for det, men det kan være nyttig å vite om at det finnes sammenhenger mellom logaritmer med ulikt grunntall. Her ser vi på sammenhengen mellom tierlogaritmen og den naturlige logaritmen, men tilsvarende sammenhenger kan bevises for logaritmer av alle grunntall.

```{admonition} Sammenheng mellom logaritmer
:class: def
Dersom $a$ er et positivt tall, så vil

$\lg a = \frac{\ln a}{\ln 10}$
```

```{admonition} Bevis for sammenhengen
:class: def, dropdown
Vi starter med å ta utgangspuntk i definisjonen for tierlogaritmen. Deretter tar vi den naturlige logaritmen på begge sider før vi bruker tredje logaritmesetning og så løser for $\lg a$.

$\begin{align}
10^{\lg a} &= a \\
\ln 10^{\lg a} &= \ln a \\
\lg a \ln 10 &= \ln a \\
\lg a &= \frac{\ln a}{\ln 10}
\end{align}
$
```
