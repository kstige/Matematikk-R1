Fra 1T husker vi at vi jobbet med [rasjonale funksjoner](https://kstige.github.io/Matematikk-1T/docs/funksjoner/rasjonalefunksjoner.html) og asymptoter. Et eksempel på en rasjonal funksjon er $f(x)=\frac{2x+1}{x-3}$ som er tegnet under. I forbindelse med rasjonale funksjoner arbeidet vi med å finne vertikale og horisontale asymptoter. Det skal vi se mer på på neste side, men vi repeterer kort her

 * horisontale asymptoter - det funksjonen gikk mot når $x$ gikk mot $\pm \infty$.
 * vertikale asymptoter - der nevneren var 0 - når $x$ nærmet seg $x$-verdien for den vertikale asymptoten, så gikk funksjonsverdien mot $\pm \infty$.

I funksjonen over ser vi at den horisontale asymptoten er $y=2$. En matematisk skrivemåte for dette er at

$$\lim_{x\to \pm\infty}f(x)=2$$

Vi leser det som at grenseverdien til $f(x)$ når $x$ går mot pluss/minus uendelig er 2.

Tilsvarende har vi en vertikal asymptote ved $x=3$. Da kan vi skrive

$$\lim_{x\to 3}f(x)=\pm \infty$$

Det leser vi som at grenseverdien til $f(x)$ når $x$ går mot 3 er pluss/minus uendelig. At grenseverdien går mot uendelig (og i tillegg med begge fortegn) skal vi senere definere som at grenseverdien ikke eksisterer.

## Grenseverdier
Over så vi to eksempler på det vi i matematikk kaller grenseverdier. 

```{admonition} Grenseverdi
:class: def
La $f$ være en funksjon. Dersom $f(x)$ går mot en verdi $b$ når $x$ går mot $a$, så sier vi at $b$ er grenseverdien til $f$ når $x$ går mot $a$. Vi skriver grenseverdien som

$\displaystyle\lim_{x\to a} f(x)=b$

Funksjonen trenger ikke være definert for $x=a$ for at grenseverdien skal eksistere.

Dersom funksjonen er definert for $x=a$ vil 

$\displaystyle\lim_{x\to a} f(x)=f(a)$
```

```{admonition} Eksempel: Grenseverdi for andregradsfunksjon
:class: eksempel

Vi ser på andregradsfunksjonen $f(x)=x^2+2x-3$ og skal finne grenseverdien når $x\to 3$. Definisjonsmengden til funksjonen er alle reelle tall. Da vil 

$\displaystyle\lim_{x\to 3} f(x)=3^2+2\cdot 3-3=12=f(3)$

Resultatet betyr at når $x$ nærmer seg 3, så vil $f$ nærme seg 12. Vi kan komme så nærme 12 vi vil ved å la $x$ gå nærme nok 3.
```

I eksempelet over kunne vi bare sette inn for $x$ for å finne grenseverdien. Slike eksempel er stort sett kun interessante fra et teoretisk og innlæringsperspektiv. I praksis vil vi være interessert i grenseverdier for $x$-verdier der funksjonen ikke er definert.

`````{admonition} Eksempel: Grenseverdi for rasjonal funksjon
:class: eksempel

$f(3)=\frac{2(3^2-9)}{3-3}=\frac{0}{0}$

som ikke gir noen mening. Hva er det som skjer her?

For å svare på det må vi bruke grenseverdier. Vi starter først med å finne grenseverdien i CAS.

Vi ser at grenseverdien til funksjonen er 12. Matematisk skriver vi det som at

$\displaystyle\lim_{x\to 3} f(x)= 12$

Det betyr at når $x$ kommer nærmere og nærmere 3, så vil $f$ komme nærmere og nærmere 12. Vi kan aldri sette inn $x=3$, men vi kan komme så nært vi vil. Så akkurat i punktet $(3,12)$ har vi et hull på grafen vår, men punktet $(2.999, 11.998)$ er der (og tilsvarende punkt enda nærmere). Vi kan markere det her på grafen med et lite kryss.

`````

Over fant vi grenseverdien ved å se på grafen og bruke CAS. Det vil også være nyttig å kunne finne grenseverdier uten hjelpemidler siden det vil hjelpe oss blant annet med å forstå funksjoner som i eksempelet over. Grenseverdier blir også viktig for å få bedre forståelse for derivasjon som er et av de viktigste temaene i R1.

For å arbeide med grenseverdier trenger vi noen regneregler. Disse er i utgangspunktet ganske intuitive (selv om det kan være litt kronglete å bevise det), så i praksis vil vi sjelden tenke over at vi bruker noen bestemte regler.

```{admonition} Regneregler for grenseverdier
:class: def

$\displaystyle\lim_{x\to a} x = a$

$\displaystyle\lim_{x\to a} k = k$, der $k\in\mathbf{R}$. Grenseverdien til et tall er tallet.

Videre antar vi at $f$ og $g$ er to funksjoner og at grenseverdiene $\displaystyle\lim_{x\to a}f(x)$ og $\displaystyle\lim_{x\to a}g(x)$ eksisterer.

$\displaystyle\lim_{x\to a}\left(f(x)\pm g(x)\right)=\lim_{x\to a}f(x)\pm\lim_{x\to a}g(x)$

$\displaystyle\lim_{x\to a}\left(f(x)\cdot g(x)\right)=\lim_{x\to a}f(x)\cdot\lim_{x\to a}g(x)$

$\displaystyle\lim_{x\to a}\frac{f(x)}{g(x)}=\frac{\lim_{x\to a}f(x)}{\lim_{x\to a}g(x)}$ når $\lim_{x\to a}g(x)\neq 0$
```

For å illustrere regnereglene tar vi først et enkelt eksempel før vi prøver på den rasjonale funksjonen i forrige eksempel.

```{admonition} Eksempel: bruke regnereglene
:class: eksempel

$\begin{align}
\displaystyle\lim_{x\to 3}\frac{2x+3}{3x}&=\displaystyle\frac{\lim_{x\to 3}(2x+3)}{\lim_{x \to 3}3x}\\
&=\displaystyle\frac{\lim_{x\to 3}2\cdot\lim_{x\to 3}x+\lim_{x\to 3}3}{\lim_{x\to 3}3\cdot\lim_{x\to 3}x}\\
&=\displaystyle\frac{2\cdot 3+3}{3\cdot 3}\\
&=\displaystyle\frac{9}{9}=1
\end{align}$

Her har vi gjort mange unødvendige mellomregninger for å vise hvordan regnereglene i praksis brukes. Normalt vil vi føre som under.

$\displaystyle\lim_{x\to 3}\frac{2x+3}{3x}=\frac{2\cdot 3+3}{3\cdot 3}=\frac{9}{9}=1$

I dette eksempelet var $x$ i definisjonsmengden til uttrykket vi så på, så i praksis er dette bare det samme som å sette inn 3 for $x$.
```

```{admonition} Eksempel: $\frac{0}{0}$-uttrykk
:class: eksempel

Vi går nå tilbake til funksjonen $f(x)=\frac{2(x^2-9)}{x-3}$ og skal se på $\displaystyle\lim_{x\to 3} f(x)$. Om vi hadde prøvd på framgangsmåten i forrige eksempel ville vi kommet til $\displaystyle\lim_{x\to 3} f(x)=\frac{0}{0}$. Det betyr at $x=3$ er et nullpunkt både i teller og nevner. Hadde det kun vært et nullpunkt i nevner ville vi fått en vertikal asymptote og grenseverdien ville blitt uendelig (ikke eksistert). Vi har imidlertid allerede sett at den skal bli 12 og må derfor anvende litt av det vi lærte om faktorisering i 1T for å komme videre.

$\displaystyle\lim_{x\to 3} f(x)=\lim_{x\to 3} \frac{2(x^2-9)}{x-3} = \lim_{x\to 3}\frac{2(x-3)(x+3)}{x-3}$

Uttrykket er ikke definert når $x=3$. Vi kan derfor sette $x\neq 3$ slik at $x-3\neq 0$ og da forkorte vekk den felles faktoren $x-3$.

$\displaystyle\lim_{x\to 3}\frac{2(x-3)(x+3)}{x-3}=\lim_{x\to 3} 2(x+3)=\lim_{x\to 3} (2x+6)=2\cdot 3+6=12$

I den siste utrekninga kan det se ut som vi motsier oss selv litt ved å sette inn $x=3$ etter å ha sagt at $x\neq 3$. Det vi imidlertid har gjort er å bruke regnereglene for grenseverdier til å si at det her er grenseverdien når $x$ nærmer seg 3.
```

I eksempelet ser vi at $\frac{2(x^2-9)}{x-3}$ forkortes til $2x+6$. Det vil si at funksjonen 

$$f(x)=\frac{2(x^2-9)}{x-3}$$

er lik funksjonen

$$g(x)=2x+6,\quad D_g=\mathbf{R}\backslash \{3\}$$

Det forklarer hvorfor funksjonsgrafen ser ut som grafen til ei rett linje. Denne metoden med å forkorte vekk felles faktor vil vi ofte få bruk for når vi møter på $\frac{0}{0}$-uttrykk.

```{admonition} $\frac{0}{0}$-uttrykk
:class: def
La funksjonen $f(x)$ være en rasjonal funksjon gitt ved $f(x)=\frac{t(x)}{n(x)}$ der $t(x)$ og $n(x)$ er polynomfunksjoner. Dersom $n(a)=0$ vil ikke funskjonen $f$ være definert for $x=a$. I det punktet vil vi vanligvis ha en vertikal asymptote og grenseverdien $x\to a$ vil ikke eksistere. Om vi samtidig har $t(a)=0$ kan vi prøve å faktorisere ut en felles faktor $x-a$ som vi forkorter vekk. Om den nye nevneren ikke lenger har nullpunkt for $x=a$, så vil grenseverdien eksistere.
```

```{admonition} Eksempel: null i nevner
:class: eksempel
Her ser vi på grenseverdien når $x\to 1$ av to ganske like uttrykk.

$\displaystyle\lim_{x\to 1}\frac{x^2-2}{x-1}$ eksisterer ikke siden nevner blir 0, mens teller er ulik 0.

$\displaystyle\lim{x\to 1}\frac{x^2-1}{x-1}=\lim{x\to 1}\frac{(x-1)(x+1)}{x-1}=\lim{x\to 1}(x+1)=1+1=2$

Den siste grenseverdien eksisterer siden teller og nevner har et felles nullpunkt som vi kan forkorte vekk. Det er imidlertid viktig å merke seg at uttrykket allikevel ikke har en verdi for $x=1$, men at verdien til uttrykket nærmer seg 2 når $x$ nærmer seg 1.
```

```{admonition} Eksempel: Vanskeligere uttrykk
:class: eksempel

$\displaystyle\lim_{x\to 9}\frac{3-\sqrt{x}}{9-x}=\lim_{x\to 9}\frac{3-\sqrt{x}}{(3-\sqrt{x})(3+\sqrt{x})}=\lim_{x\to 9}\frac{1}{3+\sqrt{x}}=\frac{1}{3+\sqrt{9}}=\frac{1}{6}$
```

## Ensidige grenseverdier

I noen tilfeller vil vi komme til å bruke det vi kaller ensidige grenseverdier.

```{admonition} Ensidige grenseverdier
:class: def
Grenseverdien til $f(x)$ når $x$ nærmer seg $a$ nedenfra (fra venstre) skrives

$$\lim_{x\to a^-} f(x)$$

Grenseverdien til $f(x)$ når $x$ nærmer seg $a$ ovenfra (fra høyre) skrives

$$\lim_{x\to a^+} f(x)$$
```

Et viktig resultat om ensidige grenseverdier er

```{admonition} Eksistens av grenseverdi
:class: def
Dersom $\lim_{x\to a^-} f(x)= \lim_{x\to a^+} f(x)$, så vil grenseverdien $\lim_{x\to a} f(x)$ eksistere og være lik verdien til de ensidige grenseverdiene.

Dersom $\lim_{x\to a^-} f(x) \neq \lim_{x\to a^+} f(x)$, så vil grenseverdien $\lim_{x\to a} f(x)$ ikke eksistere.
```

Vi ser på noen eksempler på det.

```{admonition} Eksempel: Rasjonal funksjon
:class: eksempel
Vi ser på funksjonen $f(x)=\frac{2x+1}{x+2}$ og vil se på de ensidige grenseverdiene når $x$ går mot -2.

$\displaystyle\lim_{x\to -2^-} f(x) = \infty$ siden telleren er negativ og nevneren går mot null fra negativ side.

$\displaystyle\lim_{x\to -2^+} f(x) = -\infty$ siden telleren er negativ og nevneren går mot null fra positiv side.

Siden $\displaystyle\lim_{x\to -2^-} f(x) \neq \displaystyle\lim_{x\to -2^+} f(x)$, så eksisterer ikke grenseverdien $\displaystyle\lim_{x\to -2} f(x)$.
```

`````{admonition} Eksempel: Logaritmisk funksjon
:class: eksempel
Vi ser på funksjonen $f(x)=\ln(x-3)$ og grenseverdien når $x\to 3$. Fra definisjonene av logaritmer husker vi at vi kun ta logaritmen av positive tall. Her er det derfor umulig å la $x\to 3^-$ siden vi ikke har noen funksjonsverdier på den negative siden av 3. Vi kan derfor skrive

$\displaystyle\lim_{x\to 3^-} f(x)$ eksisterer ikke.

Dersom vi ser på grafen til funksjonen kan det se ut som at funksjonen har en vertikal asymptote for $x=3$.

```{figure} ./bilder/lnx-3.png
---
scale: 40%
---
```

Vi konkluderer derfor med at 

$\displaystyle\lim_{x\to 3^+} f(x)=-\infty$
`````

Resultatet i det siste eksempelet kan uttrykkes generelt.

```{admonition} Grenseverdier i logaritmer
:class: def
For funksjonen $f(x)=\ln(x-a)$ vil 

$\displaystyle\lim_{x\to a^-} f(x)$ eksisterer ikke.

$\displaystyle\lim_{x\to a^+} f(x)=-\infty$
```

## Grenseverdier når $x\to\pm\infty$
Til nå har vi sett på grenseverdier der $x$ går mot et bestemt tall. I tillegg vil vi ofte være interessert i grenseverdien når $x\to\pm\infty$. Den sier oss hva som skjer med funksjonen når $x$ blir veldig stor. Vil funksjonen vokse mot uendelig, eller vil den gå mot en bestemt verdi. Det første er tilfellet for funksjoner som $f(x)=x^2$, mens den rasjonale funksjonen vi så på øverst på siden nærmet seg den horisontale asymptoten $y=2$.

Vi oppsummerer først noen resultater (uten bevis) som forhåpentligvis er ganske intuitive.

```{admonition} Grenseverdier for polynomfunksjoner
:class: def
Alle polynomfunksjoner vil gå mot enten positiv eller negativ uendelig ($\pm\infty$) når $x\to\pm\infty$.
```

```{admonition} Grenseverdier for eksponentialfunksjoner
:class: def
Vi lar $f(x)=a^x$.

**$0<a<1$**\
$\displaystyle\lim_{x\to\infty} a^x =0$\
$\displaystyle\lim_{x\to -\infty} a^x =\infty \quad \text{grenseverdien eksisterer ikke}$

**$a=1$**\
$\displaystyle\lim_{x\to\infty}=1^x=1$\
$\displaystyle\lim_{x\to -\infty}=1^x=1$

**$a>1$**\
$\displaystyle\lim_{x\to\infty}=a^x=\infty \quad \text{grenseverdien eksisterer ikke} $\
$\displaystyle\lim_{x\to -\infty}=a^x=0$

Hvis du ser for deg hvordan grafen til en eksponentialfunksjon ser ut, så bør resultatene være greie å huske. Om du ikke gjør det, så bør du tegne opp noen eksponentialfunksjoner og sammenlikne med grenseverdiene over. Alternativt kan resultatene forklares med utgangspunkt i definisjonen av en potens.
```

```{admonition} Grenseverdier for logaritmefunksjoner
:class: def
Logaritmefunksjoner vokser veldig sakte, men de vil gå mot uendelig når $x$ går mot uendelig.

$\displaystyle\lim_{x\to\infty} \ln x = \infty \quad \text{grenseverdien eksisterer ikke.}$
```

Til slutt skal vi se på rasjonale funksjoner.

```{admonition} Grenseverdier for rasjonale funksjoner
:class: def
En rasjonal funksjon er på formen $f(x)=\frac{t(x)}{n(x)}$, der $t(x)$ og $n(x)$ er polynomfunksjoner. Vi har tre mulige tilfeller når vi ser på grenseverdien i grensen $x\to\pm\infty$.

 * Dersom $t(x)$ har høyere grad enn $n(x)$, så vil telleren vokse raskere enn nevneren og $f(x)$ går mot uendelig i positiv eller negativ retning.
 * Dersom $n(x)$ har lavere grad enn $t(x)$, så vil nevneren vokse raskere enn telleren og $f(x)$ går mot null.
 * Dersom $t(x)$ og $n(x)$ har samme grad vil grenseverdien gå mot et tall gitt ved forholdet mellom de leddene med høyest grad.
```

Et triks vi kan bruke for å se alt dette lettere er å dele alle ledd i det rasjonale uttrykket på den høyeste potensen av $x$.

```{admonition} Eksempel 1: Grenseverdi for rasjonale funksjoner
:class: eksempel
$\displaystyle\lim_{x\to\infty}\frac{x^2+2x-4}{2x^3-3x^2+5}=0$ siden nevneren har høyere grad enn telleren. 

Det blir kanskje lettere å se det dersom vi deler alle ledd i funksjonen på den høyeste potensen av $x$ som er $x^3$.

$\begin{align}
\displaystyle\lim_{x\to\infty}\frac{x^2+2x-4}{2x^3-3x^2+5}&=\lim_{x\to\infty}\frac{\frac{x^2}{x^3}+\frac{2x}{x^3}-\frac{4}{x^3}}{\frac{2x^3}{x^3}-\frac{3x^2}{x^3}+\frac{5}{x^3}} \\
&=\lim_{x\to\infty}\frac{\frac{1}{x}+\frac{2}{x^2}-\frac{4}{x^3}}{2-\frac{3}{x}+\frac{5}{x^3}}\\
&=\frac{0+0-0}{2+0+0}=0
\end{align}$
```

```{admonition} Eksempel 2: Grenseverdi for rasjonale funksjoner
:class: eksempel
$\displaystyle\lim_{x\to\infty}\frac{x^3+9}{4x^2-2x}=\infty$ siden telleren har høyere grad enn nevneren. 

Det blir kanskje lettere å se det dersom vi deler alle ledd i funksjonen på den høyeste potensen av $x$ som er $x^3$.

$\begin{align}
\displaystyle\lim_{x\to\infty}\frac{x^3+9}{4x^2-2x}&=\lim_{x\to\infty}\frac{\frac{x^3}{x^3}+\frac{9}{x^3}}{\frac{4x^2}{x^3}-\frac{2x}{x^3}} \\
&=\lim_{x\to\infty}\frac{1+\frac{9}{x^3}}{\frac{4}{x}-\frac{2}{x^2}}\\
&=\frac{1+0}{0-0}=\infty
\end{align}$

siden vi prøver å dele på null.
```

```{admonition} Eksempel 3: Grenseverdi for rasjonale funksjoner
:class: eksempel
$\displaystyle\lim_{x\to\infty}\frac{3x^2+4}{5x^2-2x+1}=\frac{3}{5}$. 

Det blir kanskje lettere å se det dersom vi deler alle ledd i funksjonen på den høyeste potensen av $x$ som er $x^2$.

$\begin{align}
\displaystyle\lim_{x\to\infty}\frac{3x^2+4}{5x^2-2x+1}&=\lim_{x\to\infty}\frac{\frac{3x^2}{x^2}+\frac{4}{x^2}}{\frac{5x^2}{x^2}-\frac{2x}{x^2}+\frac{1}{x^2}} \\
&=\lim_{x\to\infty}\frac{3+\frac{4}{x^2}}{5-\frac{2}{x}+\frac{1}{x^2}}\\
&=\frac{3+0}{5-0+0} \\
&=\frac{3}{5}
\end{align}$
```
